```python
# import the library 
import pandas as pd

# load the dummy Dataset
df_covid19 = pd.read_excel('COVID-19-geographic-disbtribution-worldwide-2020-08-20.xlsx')
```

```python
# groupby the dataset with countriesAndTerritories
# and get the sum of individual cases and deaths for each countries.

df_covid19.groupby("countriesAndTerritories")[['countriesAndTerritories','cases','deaths']].sum()
```
