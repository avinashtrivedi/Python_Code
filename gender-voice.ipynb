```python
# import the library
import pandas as pd
```

```python
# load the Dataset and create train and test DataFrame  
df_train = pd.read_csv('gender-voice-train.csv')
df_test = pd.read_csv('gender-voice-test.csv')
```

```python
# number of rows and columns in test DataFrame
row, col = df_test.shape
print('Row: ',row)
print('Col: ',col)
```

```python
# Check for the missing values in the training DataFrame
# False means there is no missing values

df_train.isnull().values.any()
```

```python
# Check for the missing values in the test DataFrame
# False means there is no missing values

df_test.isnull().values.any()
```

```python
# Count the number of male and female classes in the train DataFrame
# In the label column, 0 denotes the male class and 1 denotes the female class.

df_train['label'].value_counts()
```

```python
# Separate the feature variables
X_train = df_train.iloc[:,1:].values 
X_test = df_test.iloc[:,1:].values 
```

```python
# Separate the target variable
y_train = df_train.iloc[:,0].values
y_test  = df_test.iloc[:,0].values
```

```python
# Fit the Random Forest Classifier to the Training set
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(random_state = 42)
classifier.fit(X_train, y_train)
```

```python
# prediction on Test set
y_pred = classifier.predict(X_test)
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# classification report 
print(classification_report(y_test, y_pred))
```

```python
# accuracy
acc = accuracy_score(y_test, y_pred)
```

```python
# round() : round the values 10 3 decimal places 
print('accuracy: ',round(acc,3))
```

- Based on the confusion matrix, precision, recall ,f1-score and accuracy score , it is observed that the prediction model is making accurate prediction 97.8% of the time.
