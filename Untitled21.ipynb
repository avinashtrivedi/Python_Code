```python
# Pokemon lists
pokemon_names = ['charmander', 'squirtle', 'bulbasaur', 'gyrados']
pokemon_amounts = [3, 2, 5, 1]
adoption_fees = [100.00, 50.00, 25.00, 1000.00]
pokemon_types = [['fire'], ['water'], ['grass'], ['water', 'flying']]
valid_pokemon_types = ['bug', 'dark', 'dragon', 'electric', 'fairy', 'fighting', 'fire', 'flying', 'ghost', 'grass', 'ground', 'ice', 'normal', 'poison', 'psychic', 'rock', 'steel', 'water']
flag = 0

# lowercase all the pokemons in pokemon_names
[x.lower() for x in pokemon_names]

# Loop until user chooses to quit
while True:
    # welcome message
    print('Welcome to the Pokemon Center!')

    # input whether user wants to search or quit
    print('(a)dd, (r)emove, r(e)port, (s)earch by name, (t)type, (l)list or (q)uit: ', end="")
    user_input = input()

    # if user chooses to quit, break the loop
    if user_input == "q" or user_input == "Q":
        print('See you next time!')
        break

    # if user chooses to see list
    elif user_input == 'l' or user_input == 'L':
        # print formatted list
        print('Name         Amount Available   Adoption Fee  Type(s)')
        for i in range(len(pokemon_names)):
            print('%-13s' % (pokemon_names[i].capitalize()), end="")
            print('%16d %14.2f  ' % (pokemon_amounts[i], adoption_fees[i]), end="")
            for type in pokemon_types[i]:
                print(type.capitalize(), end=" ")
            print()
        print()

    # if user chooses to search
    elif user_input == "s" or user_input == "S":
        # get pokemon name
        pokemon = input("Name of Pokemon to search for: ")

        # get the index at which pokemon is present in pokemon_names
        index = -1
        for i in range(len(pokemon_names)):
            if pokemon_names[i] == pokemon.lower():
                index = i
                break

        # capitalize first letter of string
        pokemon = pokemon.lower()
        pokemon = pokemon.capitalize()

        # if no pokemon of that name exist
        if index == -1:
            print('We do not have any', pokemon, 'at the Pokemon Center')
            print()
            continue

        # print result
        print('We have', pokemon_amounts[index], pokemon, 'at the Pokemon Center')
        print('It will cost $%.2f' % (adoption_fees[index]), 'to adopt this Pokemon')
        print(pokemon, 'has the following types: ', end="")
        for type in pokemon_types[index]:
            print(type.capitalize(), end=" ")
        print()
        print()

    # Search by type 
    elif user_input == 't' or user_input == 'T':

        # Get type input 
        pokemon_type = input('Enter Pokemon type: ')
        pokemon_type = pokemon_type.lower()
        for types in pokemon_types:
              if pokemon_type in types:
                  flag = 1
        if flag != 1:
              print('We have no Pokemon of type',pokemon_type,'at our Pokemon Center')
              print()
              flag = 0
              continue
        # print formatted list
        print('Name         Amount Available   Adoption Fee  Type(s)')
        for i in range(len(pokemon_names)):
            if pokemon_type in pokemon_types[i]:
                print('%-13s' % (pokemon_names[i].capitalize()), end="")
                print('%16d %14.2f  ' % (pokemon_amounts[i], adoption_fees[i]), end="")
                for type in pokemon_types[i]:
                    print(type.capitalize(), end=" ")
                print()
        print()
        flag = 0 

    #Add pokemon
    elif user_input == 'a' or user_input == 'A':
        temp_type = []
        #Name
        pokemon_name = input('Enter name of new pokemon: ')
        if pokemon_name in pokemon_names:
            print('Duplicate name, add operation cancelled')
            continue
        else:
            pokemon_names.append(pokemon_name)

            #Count
            number = int(input('How many of these Pokemon are you adding? '))
            while number <= 0:
                print('Invalid, please try again')
                number = int(input('How many of these Pokemon are you adding? '))
            pokemon_amounts.append(number)

            #fees
            fees = float(input('What is the adoption fee for this Pokemon? '))
            while fees <= 0:
                print('Invalid, please try again')
                fees = float(input('What is the adoption fee for this Pokemon? '))
            adoption_fees.append(fees)

            #Type
            print('Next you will be prompted to enter the types for this Pokemon. Pokemon can have multiple types. Type help to view all possible Pokemon types, and type end to stop entering types. You must enter at least one valid type')
            input_type = input('What type of Pokemon is this? ')
            while input_type.lower() != 'end':
                if input_type.lower() == 'help':
                    for i in valid_pokemon_types:
                        print('*',i)
                else:
                    if input_type.lower() in valid_pokemon_types:
                        if input_type.lower() not in temp_type:
                            print('Type',input_type.lower(),'applied')
                            temp_type.append(input_type.lower())
                        else:
                            print('Duplicate type')
                    else:
                        print('This is not a valid type, please try again')
                input_type = input('What type of Pokemon is this? ')
            print('Pokemon added!')
            print()
            pokemon_types.append(temp_type)
            temp_type = []

    #Remove
    elif user_input == 'r' or user_input == 'R':
        # input
        pokemon_name = input('Enter name of Pokemon to remove: ')
        if pokemon_name not in pokemon_names:
            print('Pokemon not found, cannot remove')
            print()
            continue
        else:
            #If pokemon exists removing from all the lists
            position = pokemon_names.index(pokemon_name)
            pokemon_names.pop(position)
            pokemon_amounts.pop(position)
            adoption_fees.pop(position)
            pokemon_types.pop(position)
            print('Pokemon removed')
            print()
    #Report
    elif user_input == 'e' or user_input == 'E':
        highest_index = adoption_fees.index(max(adoption_fees))
        lowest_index = adoption_fees.index(min(adoption_fees))
        total_cost = 0
        for i in range(len(pokemon_names)):
            total_cost += adoption_fees[i]*pokemon_amounts[i]
        print('Highest priced Pokemon:',pokemon_names[highest_index],'@$',adoption_fees[highest_index],'per pokemon')
        print('Lowest priced Pokemon:',pokemon_names[lowest_index],'@$',adoption_fees[lowest_index],'per pokemon')
        print('Total cost to adopt all Pokemon in the Center: $',total_cost)
        print()
    # if invalid command is entered
    else:
        print('Unknown command, please try again')
        print()
        continue
```
