```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('gapminder.csv')
a = df.groupby(['country']).apply(lambda x:x)
```

```python
c={'Asia':'blue',
   'Africa':'green',
   'Europe':'yellow',
   'Americas':'orange',
   'Oceania':'deeppink'
  }

from matplotlib.colors import LogNorm
fgrid = sns.relplot(data=a,
                   x='year',
                   y='gdpPercap',
                   col='continent',
                   hue="continent",
                   kind='line',
                   col_wrap=3,
#                     style='continent',
#                  palette=['b','darkorange','g','r','m'],
                 size= "country",
                 legend=True,
                    ci= None,
#                  size_norm=LogNorm(),
#                  facet_kws=dict(legend_out=True)
                   )


for ax in fgrid.axes.flatten():
    ax.semilogy()
```
