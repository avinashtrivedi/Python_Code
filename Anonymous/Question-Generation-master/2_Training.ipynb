```python
# import
import pandas as pd
import numpy as np
```

### Pickling : to save the Trained Model for future reuse

```python
import _pickle as cPickle
from pathlib import Path

def dumpPickle(fileName, content):
    pickleFile = open(fileName, 'wb')
    cPickle.dump(content, pickleFile, -1)
    pickleFile.close()

def loadPickle(fileName):    
    file = open(fileName, 'rb')
    content = cPickle.load(file)
    file.close()
    
    return content
    
def pickleExists(fileName):
    file = Path(fileName)
    
    if file.is_file():
        return True
    
    return False
```

## Reading the dataframe: Generated from the preprocessing notebook

```python
wordPickleName = 'output/wordsDf.pkl'
df = loadPickle(wordPickleName)

df.head()
```

## One-hot encoding
encode the categorical data - NER, POS, TAG, DEP, shape

```python
columnsToEncode = ['NER', 'POS', "TAG", 'DEP']

for column in columnsToEncode:
    print(column)
    one_hot = pd.get_dummies(df[column])
    one_hot = one_hot.add_prefix(column + '_')

    df = df.drop(column, axis = 1)
    df = df.join(one_hot)
```

```python
df.head()
```

## Remove columns
We need to remove the columns that are not features of the words.

```python
columnsToDrop = ['text', 'titleId', 'paragrapghId', 'sentenceId', 'shape']
df = df.drop(columnsToDrop, axis = 1)
df.head()
```

## Split dataset to train and test set

```python
from sklearn.model_selection import train_test_split

x_data = df.drop(labels=['isAnswer'], axis=1)
y_data = df['isAnswer']

x_train, x_test, y_train, y_test = train_test_split(x_data, 
                                                    y_data, 
                                                    test_size=0.1, 
                                                    random_state=42)
```

```python
# train/test shape
print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(y_test))
```

## Build and Train the model

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

predictor = gnb.fit(x_train, y_train)
y_pred = predictor.predict(x_test)
```

### Show Accuracy

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print("accuracy:",round(accuracy_score(y_test, y_pred)*100,2),'%')
```

### Confusion matrix

```python
confusion_matrix(y_test, y_pred)
```

### Prettier confusion matrix: Visualization

```python
import itertools
import matplotlib.pyplot as plt

def plot_confusion_matrix(y_true, y_pred, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues,
                          figsize=(9, 7)):
    matrix = confusion_matrix(y_true, y_pred)

    if normalize:
        matrix = matrix.astype('float') / matrix.sum(axis=1)[:, np.newaxis]

    plt.figure(figsize=figsize)
    plt.imshow(matrix, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()

    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = matrix.max() / 2.
    for i, j in itertools.product(range(matrix.shape[0]), range(matrix.shape[1])):
        plt.text(j, i, format(matrix[i, j], fmt),
                 horizontalalignment="center",
                 size=int((figsize[0] / 10) * 38),
                 color="white" if matrix[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
plot_confusion_matrix(y_test, y_pred, ['not correct answer', 'correct answer'], figsize=(5, 4))
```

## Saving the model
save the predictor, so we can use it to label new words.

```python
predictorPickleName = 'output/nb-predictor.pkl'
dumpPickle(predictorPickleName, predictor)

predictorFeaturesName = 'output/nb-predictor-features.pkl'
dumpPickle(predictorFeaturesName, list(df.columns))
```
