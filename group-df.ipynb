```python
# import library
import pandas as pd
```

```python
# read the csv file
# put csv file in your current directory,i.e. where you have saved your python program
df = pd.read_csv('phonedata.csv')
```

```python
df # displaying dataframe
```

```python
# grouping on network_type and item and getting count for all remaining columns
df1 = df.groupby(['network_type','item']).count() 
df1
```

```python
# grouping on network_type and item and getting count for only duration column
df2 = df.groupby(['network_type','item'])['duration'].count().reset_index()
df2
```
