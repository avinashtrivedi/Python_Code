```python
my_dict = {'a':15, 'c':18 , 'b': 20 }
```

# Q.a  
- print all the keys

```python
# get dictionary keys using my_dict.keys()
for i in my_dict.keys():
    print(i)
```

# Q.b
- print all the values

```python
# get dictionary values using my_dict.values()
for i in my_dict.values():
    print(i)
```

# Q.c
- print all the keys and values pairs.

```python
print('{:<5} {:<5}'.format('Key','Value'))

# get dictionary key, value pairs using my_dict.items()
for k,v in my_dict.items():
    print('{:<5} {:<5}'.format(k,v))
```

# Q.d
- print all the keys and values pairs in ascending order of key

```python
print('{:<5} {:<5}'.format('Key','Value'))

# sort the dictionary on the basis of keys
for k in sorted(my_dict):
    print('{:<5} {:<5}'.format(k,my_dict[k]))
```

# Q.e
- print all the keys and values pairs in ascending order of value

```python
print('{:<5} {:<5}'.format('Key','Value'))

# sort the dictionary on the basis of values.
for k,v in sorted(my_dict.items(),key=lambda x:x[1]):
    print('{:<5} {:<5}'.format(k,v))
```
