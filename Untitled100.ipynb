```python
# load the library
import pandas as pd

# reading the data from UIC2016MensBasketBall.csv
df2016 = pd.read_csv("UIC2016MensBasketBall.csv")
```

```python
df2016.dtypes # check the datatypes
```

```python
# change datatype of Date to datetime
df2016['Date'] =pd.to_datetime(df2016['Date'])
```

```python
# Game played in 2016
newDF = df2016[df2016['Date'].dt.year==2016]
newDF
```

```python
newDF['Opp'] # get opponents name
```

```python
# get number of unique opponents
len(newDF['Opp'].unique())
```
