```python
# store the notes denomination
notes = [100,50,20,10,5,2,1]
notes_dict = {}
while True:
    amount = input('Enter Amount to be paid in $: ')
    
    # loop to prompt for valid integer Amount
    while True:
        if amount.isnumeric():
            amount = int(amount)
            break
        else:
            amount = input('This is not a valid positive integer value. Please re-enter a valid positive integer: ')
    
    # get the quantity of notes of different denomination
    for note in notes:
        qty = amount//note
        amount = amount%note
        notes_dict[note] = qty
    
    # print the result
    for key,val in notes_dict.items():
        print('denomination $: {:<5} Number of Notes {:<5}'.format(key,val))
    
    # Ask for choices, Would you like to continue?
    # loop to keep prompting for valid choices
    while True:
        choice = input('Would you like to continue? (yes/no): ')
        print()
        if choice=='no':
            flag = False
            break
        elif choice=='yes':
            flag = True
            break
        else:
            print('Wrong choice. Please try again.')
    if flag==False:
        break
```
