```python
# import the library
from PIL import ImageDraw,Image
import numpy as np

def crop_circle(fname):
    
    # try-except block to handle exception.
    try:
        # Open the input image
        oldimg = Image.open(fname)

        # convert image to numpy array
        np_oldimg = np.array(oldimg)

        # get the height and width
        height = oldimg.size[0]
        width = oldimg.size[1]

        # Create grayscale image of same size with circle
        new_img = Image.new('L', oldimg.size)
        draw = ImageDraw.Draw(new_img)

        # get the circular part of the image in white
        draw.pieslice([0,0,height,width],0,360,fill=255)

        # Convert new_img Image to numpy array
        np_new_img = np.array(new_img)

        # superimpose original image and the new image.
        np_new_img = np.dstack((np_oldimg,np_new_img))

        # Save the new image
        Image.fromarray(np_new_img).save('new_cat.png')
        print('cropped image has been created with name','new_cat.png','in current working directory.')
    except:
        print('Wrong filepath.')
```

```python
crop_circle('mycat.jpg')
```
