```python
import pandas as pd
import matplotlib.pyplot as plt

fram_tr = pd.read_csv("framingham_nb_training1.csv")
table_fram_tr_death_sex_plot = pd.crosstab(fram_tr['Sex'], fram_tr['Death'])

# create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

# plot on first subplot i.e. axes[0]
table_fram_tr_death_sex_plot.plot(kind='bar', stacked = True,ax=axes[0])
table_fram_tr_death_educ_plot = pd.crosstab(fram_tr['Educ'], fram_tr['Death'])

# plot on Second subplot i.e. axes[1]
table_fram_tr_death_educ_plot.plot(kind='bar', stacked = True,ax=axes[1])

# Adjust the padding between and around subplots
fig.tight_layout()
```
