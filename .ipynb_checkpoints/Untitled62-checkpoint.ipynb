```python
import pygame, random
# Define some colors
BLACK = (0,0,0)
WHITE = (255,255,255)
GREEN = (0,255,0)
RED = (255,0,0)
YELLOW = (255,255,0)
GREY = (100,100,100)
BLUE = (0,0,255)
background=pygame.image.load("23.png")
pygame.init()
#3 created the class
class Ninja(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.image.load("23.png").convert_alpha()
    # Set our transparent color
    #self.image.set_colorkey(WHITE)
        self.rect=self.image.get_rect()
    def update(self):
        self.rect.y+=1
        if self.rect.y>500:
            self.rect.y=-10
            self.rect.x=random.randrange(0,670)
    #9B
    def resetPos(self):
        self.rect.y=-10
        self.rect.x=random.randrange(0,670)
    
class Pom(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.image.load("23.png").convert_alpha()
        # Set our transparent color
        #self.image.set_colorkey(WHITE)
        self.rect=self.image.get_rect()
    def update(self):
        self.rect.y+=1
        if self.rect.y>500:
            self.rect.y=-10
            self.rect.x=random.randrange(0,670)
#9B
    def resetPos(self):
        self.rect.y=-10
        self.rect.x=random.randrange(0,670)
class Apple(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.image.load("23.png").convert_alpha()
        # Set our transparent color
        #self.image.set_colorkey(WHITE)

        self.rect=self.image.get_rect()
    def update(self):
        self.rect.y+=1
        if self.rect.y>500:
            self.rect.y=-10
            self.rect.x=random.randrange(0,670)
#9B
    def resetPos(self):
        self.rect.y=-10
        self.rect.x=random.randrange(0,670)
        
class Banana(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.image.load("23.png").convert_alpha()
    # Set our transparent color
    #self.image.set_colorkey(WHITE)
        self.rect=self.image.get_rect()
    def update(self):
        self.rect.y+=1
        if self.rect.y>500:
            self.rect.y=-10
            self.rect.x=random.randrange(0,670)
#9B
    def resetPos(self):
        self.rect.y=-10
        self.rect.x=random.randrange(0,670)
class Bomb(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.image.load("23.png").convert_alpha()
        # Set our transparent color
        #self.image.set_colorkey(WHITE)
        self.rect=self.image.get_rect()
    def update(self):
        self.rect.y+=1
        if self.rect.y>500:
            self.rect.y=-10
            self.rect.x=random.randrange(0,670)
#9B
    def resetPos(self):
        self.rect.y=-10
        self.rect.x=random.randrange(0,670)
        #width and height of the screen [width, height]
        size = (700, 500)
        screen = pygame.display.set_mode(size)
        pygame.display.set_caption("Fruit Ninja Final Project")
        # Loop until the user clicks the close button.
        done = False
        #9A
        # Used to manage how fast the screen updates
        clock = pygame.time.Clock()
        #1 create a groupt o hold all sprites and falling sprites
        all_sprites=pygame.sprite.Group()
        fruit=pygame.sprite.Group()
        bombs=pygame.sprite.Group()
        #4
        player=Ninja()
        all_sprites.add(player)
        #6 create 50 falling sprites
        for i in range(5):
            #r=random.randrange(0,256)
            #g=random.randrange(0,256)
            #b=random.randrange(0,256)
            apple=Apple() #create a block
            apple.rect.x=random.randrange(700-30) #generate a random x
            apple.rect.y=random.randrange(500-33) #generate a random y
            #add the block to both groups
            all_sprites.add(apple)
            fruit.add(apple)
        for i in range(5):
            #r=random.randrange(0,256)
            #g=random.randrange(0,256)
            #b=random.randrange(0,256)
            banana=Banana() #create a block
            banana.rect.x=random.randrange(700-30) #generate a random x
            banana.rect.y=random.randrange(500-46) #generate a random y
            #add the block to both groups
            all_sprites.add(banana)
            fruit.add(banana)
        for i in range(5):
            #r=random.randrange(0,256)
            #g=random.randrange(0,256)
            #b=random.randrange(0,256)
            pom=Pom() #create a block
            pom.rect.x=random.randrange(700-30) #generate a random x
            pom.rect.y=random.randrange(500-31) #generate a random y
            #add the block to both groups
            all_sprites.add(pom)
            fruit.add(pom)
        for i in range(5):
            #r=random.randrange(0,256)
            #g=random.randrange(0,256)
            #b=random.randrange(0,256)
            bomb=Bomb() #create a block
            bomb.rect.x=random.randrange(700-30) #generate a random x
            bomb.rect.y=random.randrange(500-28) #generate a random y
            #add the block to both groups
            all_sprites.add(bomb)
            bombs.add(bomb)
            score=0

while not done:
# --- Main event loop
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            done = True
    # --- Game logic should go here
    #7 update all the sprites
            fruit.update()
            bombs.update()
    #8 make a list
blocks_hit_list=pygame.sprite.spritecollide(player, fruit, False)
bombs_hit=pygame.sprite.spritecollide(player, bombs, False)
    #10
for block in blocks_hit_list:
    score+=1
    block.resetPos()
for hit in bombs_hit:
    score-=5
    hit.resetPos()
#5 player position so red square moves w mouse
    pos=pygame.mouse.get_pos()
    player.rect.x=pos[0]
    player.rect.y=pos[1]
# --- Screen-clearing code goes here
# Here, we clear the screen to white. Don't put other drawing commands
# above this, or they will be erased with this command.
# If you want a background image, replace this clear with blit'ing the
# background image.
screen.blit(background, (0,0))
#2 draw all sprites
all_sprites.draw(screen)
font=pygame.font.SysFont('Courier New', 25, True, False)
if score>=200:
    text=font.render("WINNER!!", True, WHITE)
elif score<0:
    text=font.render("GAME OVER!", True, WHITE)
else:
    text=font.render("Score: " +str(score), True, WHITE)
    screen.blit(text, [540,450])
# --- Drawing code should go here
# --- Go ahead and update the screen with what we've drawn.
pygame.display.flip()
# --- Limit to 60 frames per second
clock.tick(60)
pygame.quit
```
