```python
#Table 1, print just the department code and course number, separated by a space

def table1(courses):
    print("Table 1\n")
    
    #loop through all the courses
    for i in courses:
        
        dept_code = i[:2]
        courseNum = i[2:5]
        print(dept_code,courseNum)
```

```python
#Table 2, print all the data in the string, with a single space between each piece.

def table2(courses):
    print("Table 2\n")
    
    #loop through all the courses
    for i in courses:
        
        dept_code = i[:2]
        courseNum = i[2:5]

        # remaining string as combination of Title and Enrollments
        TitleAndEnrollments = i[5:]

        # loop through TitleAndEnrollments
        for j in range(len(TitleAndEnrollments)):

            # if character is digit/integer
            if TitleAndEnrollments[j].isdigit():

                # get title and enrollment
                title = TitleAndEnrollments[:j]
                enrollment = int(TitleAndEnrollments[j:])
                break
        
        print(dept_code,courseNum,title,enrollment)
```

```python
#Table3

def table3(courses):
    
    print("Table 3\n")
    
    #loop through all the courses
    total = 0
    for i in courses:
        dept_code = i[:2]
        courseNum = i[2:5]

        # remaining string as combination of Title and Enrollments
        TitleAndEnrollments = i[5:]

        # loop through TitleAndEnrollments
        for j in range(len(TitleAndEnrollments)):

            # if character is digit/integer
            if TitleAndEnrollments[j].isdigit():

                # get title and enrollment
                title = TitleAndEnrollments[:j]
                enrollment = int(TitleAndEnrollments[j:])
                break
                
        total = total + enrollment
        
        # title[:20] will print only first 20 characters
        print("{:<2} {:<3} {:<20} {:>3}".format(dept_code,courseNum,title[:20],enrollment))
        
    print("{:>27} {:>3}".format('Total:',total))
```

```python
#Table 4, sort the list and print

def table4(courses):
    print("Table 4 ")
    
    data = []
    #loop through all the courses
    for i in courses:
        dept_code = i[:2]
        courseNum = i[2:5]

        # remaining string as combination of Title and Enrollments
        TitleAndEnrollments = i[5:]

        # loop through TitleAndEnrollments
        for j in range(len(TitleAndEnrollments)):

            # if character is digit/integer
            if TitleAndEnrollments[j].isdigit():

                # get title and enrollment
                title = TitleAndEnrollments[:j]
                enrollment = int(TitleAndEnrollments[j:])
                break
                
        # store  dept_code,courseNum,title and enrollment
        data.append((dept_code,courseNum,title,enrollment))
    
    # sort the data on the basis of lenght of title
    data.sort(key=lambda x:len(x[2]),reverse=True)
    
    # print
    for dept_code,courseNum,title,enrollment in data:
        print("{:<2} {:<3} {:<50} {:>3}".format(dept_code,courseNum,title,enrollment))
```

```python
def main():
    courses = []
    lines = open("courses.txt","r").readlines()
    for line in lines:
        courses.append(line.strip("\n"))
        
    #print(courses)             
    #display table1
    table1(courses)
    print()
    #display table2
    table2(courses)
    print()
    #display table3
    table3(courses)
    print()
    #display table4
    table4(courses)

#call main
main()
```
