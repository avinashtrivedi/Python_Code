```python
# Necessary imports and read .json file
import json
from Unit08_vincenty import latLngDistanceMeters
fh = open('Unit08_umich_buildings.json','r')
umich_buildings_str = fh.read()
umich_buildings = json.loads(umich_buildings_str)
```

```python
umich_buildings # display the content of umich_buildings
```

```python
bldgs = [('ccrb','aep'),('taubl','nq')]
def buildingDistance(building1_id,building2_id):
    
    # Try-Except block to handle exception
    try:
        for i in umich_buildings:
            
            # get the latitude/longitude coresponding to building id
            if i['id'] == building1_id:
                loc1 = (i['lat'],i['lng'])
    
            # get the latitude/longitude coresponding to building id
            if i['id'] == building2_id:
                loc2 = (i['lat'],i['lng'])
        distance = latLngDistanceMeters(loc1,loc2) # compute the distance between both the buildings
        return distance
    except:
        return None
    
# Iterate through bldgs and compute the distance
for i in bldgs:
    print("The vincenty distance between building {} and {} is {}".format(i[0],i[1],buildingDistance(i[0],i[1])))
```

```python
fname = open("north_quad_distance.csv", "w")
fname.write("{},{}".format('Name','Distance from NQ'))
for i in umich_buildings:
    if i['id']=='nq':
        loc1 = (i['lat'],i['lng'])
        
for i in umich_buildings:
    loc2 = (i['lat'],i['lng'])
    distance = latLngDistanceMeters(loc1,loc2)
    fname.write("\n{},{}".format(i['name'],distance))
fname.close()
```
