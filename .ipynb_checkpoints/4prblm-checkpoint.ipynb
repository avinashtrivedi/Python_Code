# 4.1

```python
# Loading data:
# Import function for loading the 'boston' dataset.
from sklearn.datasets import load_boston
# Load a 'bunch' containing data and descriptions.
boston_bunch = load_boston()
# Extract and store predictor variables.
X = boston_bunch.data
# Extract and store the variable that is the target for prediction.
y = boston_bunch.target
# Extract and store labels of predictor variables.
labels = boston_bunch.feature_names
```

```python
# import the library
import pandas as pd

# get the dataframe from X for better understanding.
df = pd.DataFrame(X,columns=labels)
df
```

```python
# create a new column 'MEDV' and assign y
df['MEDV'] = y
df
```

```python
# split the dataset
# 67% for training set and 33% for test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.33, random_state = 553)
```

# 4.2

```python
# import the library for OLS and r-square
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

# model #1

```python
# build the OLS model
model_1 = LinearRegression()

# fit the data
model_1.fit(X_train,y_train)

#predictions on training data
y_pred_train_1 = model_1.predict(X_train) 

#predictions on test data
y_pred_test_1 = model_1.predict(X_test)  
```

# model #2

```python
# since you need to use only AGE, NOX, DIS, and RAD column
# so get their index i.e. 6,4,7,8
labels
```

```python
# get the training and test data 
X_train_new = X_train[:,[6,4,7,8]]
X_test_new = X_test[:,[6,4,7,8]]
```

```python
# build the OLS model and fit the data
model_2 = LinearRegression()
model_2.fit(X_train_new,y_train)
y_pred_train_2 = model_2.predict(X_train_new) #predictions on training data
y_pred_test_2 = model_2.predict(X_test_new)   #predictions on test data
```

# model# 3 

```python
from sklearn.preprocessing import PolynomialFeatures

# Generate polynomial and interaction features
poly_reg = PolynomialFeatures(degree = 2)

# Fit to data, then transform it
X_train_poly = poly_reg.fit_transform(X_train)

# build the model
model_3 = LinearRegression()

# fit the polynomial features
model_3.fit(X_train_poly, y_train)

# get predictions for training and test set
y_pred_train_3 = model_3.predict(poly_reg.fit_transform(X_train))
y_pred_test_3 = model_3.predict(poly_reg.fit_transform(X_test))
```

# 4.3

# Model 1

```python
print('The R^2 of the predictions on the training set:',r2_score(y_train, y_pred_train_1))
print('The R^2 of the predictions on the test set:',r2_score(y_test, y_pred_test_1))

print('TRUE_MEDV:',y_test[:5])
print('PRED_MEDV:',y_pred_test_1[:5])
```

# Model 2

```python
print('The R^2 of the predictions on the training set:',r2_score(y_train, y_pred_train_2))
print('The R^2 of the predictions on the test set:',r2_score(y_test, y_pred_test_2))

print('TRUE_MEDV:',y_test[:5])
print('PRED_MEDV:',y_pred_test_2[:5])
```

# Model 3

```python
print('The R^2 of the predictions on the training set:',r2_score(y_train, y_pred_train_3))
print('The R^2 of the predictions on the test set:',r2_score(y_test, y_pred_test_3))

print('TRUE_MEDV:',y_test[:5])
print('PRED_MEDV:',y_pred_test_3[:5])
```

# 4.4

- Since,higher R-square indicates a better fit for the model. Therefore based on the R-square values of three models. It is safe to assume that model 3 is the best. It has highest R-square value on training (0.9425) and test set(0.8258).
