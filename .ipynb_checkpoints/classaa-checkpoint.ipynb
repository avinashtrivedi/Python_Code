```python
import question
def main():


    print("Welcome to Trivia Game! There are 10 questions.")
    print("There are 2 players taking turns after each question.")
    print("-----------------------------------------------------")

# define variables for player 1 score and player 2 score
# initialise both to zero
    tries = 10
    player1Score = 0
    player2Score = 0

# create question list
    questions = get10Questions()

# for loop to traverse through the 10 questions
    for questionCount in range(tries):
        # if else to figure out if player 1 or player 2
        if questionCount % 2 == 0:
            player = 'first'
        else:
            player = 'second'
        
        print("This question is for the", player, "player")
        print() # Added for better formatting
        # show the next question
        current = questions[questionCount]
        print(current)

        # get the user input
        print()    # Added for better formatting
        
        # Modified here so as to handle the wrong input that is not digit ,
        # suppose by mistake user has entered some character instead of digit
        playerAnswer = input("Please enter your answer, given options 1 to 4: ")
        
        # playerAnswer.isdigit()==False means , input was string/character
        while playerAnswer.isdigit()==False or int(playerAnswer) < 1 or int(playerAnswer) > 4: # while the user enters option other than 1 to 4, ask them to enter again
            print("This is not a valid entry. Please enter either 1, 2, 3 or 4.")
            playerAnswer = input("Please enter your answer, given options 1 to 4: ")
        
        playerAnswer = int(playerAnswer)
        #print('\n') # removed
        print() # Added for better formatting

        # if the answer is correct, add a point to the player's score
        if current.rightAnswer(playerAnswer):
            print('The Answer is Correct.')
            if player == "first":
                player1Score += 1
            else:
                player2Score += 1
        # print that the answer is wrong
        else:
            print('The Answer is Wrong.')
        
        print('The Correct Answer is:',current.getCorrectAns()) # print the correct answer.
        print() # Added for better formatting
        print("-----------------------------------------------------") # Added for better formatting

    # display the players' scores
    print('Player 1 has a total score of ', player1Score, 'points')
    print('Player 2 has a total score of ', player2Score, 'points', '\n')

    # figure out who wins
    if player1Score > player2Score:
        print('Congrats! The winner is Player 1')
    elif player2Score > player1Score:
        print('Congrats! The winner is Player 2')
    else:
        print('There is a tie. No winner declared.')

def get10Questions():
    questions = []

    # create each question and add it to the list
    ques1 = question.Question("Question 1: How many days are in a lunar year?", "354", "365", "243", "379", 1)
    questions.append(ques1)

    ques2 = question.Question("Question 2: What is the largest planet?", "Mars", "Jupiter", "Earth", "Saturn", 2)
    questions.append(ques2)

    ques3 = question.Question("Question 3: What is the largest kind of whale?", "Orca whale", "Humpback whale", "Beluga whale", \
                             "Blue whale", 4)
    questions.append(ques3)

    ques4 = question.Question("Question 4: Which dinosaur could fly?", "Triceratops", "Tyrannosaurus Rex", "Pteranodon", \
                              "Diplodocus", 3)
    questions.append(ques4)

    ques5 = question.Question("Question 5: Which of these Winnie the Pooh characters is a donkey?", "Pooh", "Eeyore", "Piglet", \
                              "Kanga", 2)
    questions.append(ques5)

    ques6 = question.Question("Question 6: What is the hottest planet?", "Mars", "Pluto", "Earth", "Venus", 4)
    questions.append(ques6)

    ques7 = question.Question("Question 7: Which dinosaur had the largest brain compared to body size?", "Troodon", "Stegosaurus", \
                              "Ichthyosaurus", "Gigantoraptor", 1)
    questions.append(ques7)

    ques8 = question.Question("Question 8: What is the largest type of penguin?", "Chinstrap penguin", "Macaroni penguin", \
                              "Emperor penguin", "White-flippered penguin", 3)
    questions.append(ques8)

    ques9 = question.Question("Question 9: What children's story character is a monkey?", "Winnie the Pooh", "Curious George", \
                              "Horton", "Goofy", 2)
    questions.append(ques9)

    ques10 = question.Question("Question 10: How long is a year on Mars?", "550 Earth days", "498 Earth days", "126 Earth days", \
                               "687 Earth days", 4)
    questions.append(ques10)

    return questions

# call main
main()
```
