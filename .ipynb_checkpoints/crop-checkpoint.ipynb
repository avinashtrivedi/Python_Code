```python
# state list
STATES = ['Alaska', 'Alabama', 'Arizona', 'Arkansas', 'California', 'Colorado', 
          'Connecticut', 'Delaware', 'Florida', 'Georgia', 'Hawaii', 'Idaho', 
          'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky', 'Louisiana', 'Maine', 
          'Maryland', 'Massachusetts', 'Michigan', 'Minnesota', 'Mississippi', 'Missouri', 
          'Montana', 'Nebraska', 'Nevada', 'New Hampshire', 'New Jersey', 'New Mexico', 
          'New York', 'North Carolina', 'North Dakota', 'Ohio', 'Oklahoma', 'Oregon', 'Pennsylvania', 
          'Rhode Island', 'South Carolina', 'South Dakota', 'Tennessee', 'Texas', 'Utah', 'Vermont',
          'Virginia', 'Washington', 'West Virginia', 'Wisconsin', 'Wyoming']
```

```python
# open a file and return its object
def open_file():
    '''
    it will return file object.
    '''
    try:
        # input file name
        FileName = input('Enter a file: ')
        fp = open(FileName, "r" )
        
    except IOError:
        print( "\n Wrong file name/path \n" )
        fp = None

    return fp
```

```python
# read a file and return dictionary
def read_file(fp):
    data = fp.readlines() # read all lines
    fp.close() # close the file
    
    # split data as a list of rows like this:
    #   ['Alabama ',
    #  'Upland cotton',
    #  'Genetically engineered (GE) upland cotton',
    #  'All GE varieties',
    #  '2000',
    #  'Percent of upland cotton planted',
    #  '.']
    
    
    data = [i.strip().split(',') for i in data[1:]]
    mydict = {} 
    
    # apply the condition
    # Value should be in digit
    # consider only 'All GE varieties'
    for row in data:
        if row[6].isdigit() == True and row[3]=='All GE varieties':
            crop = row[1]
            if 'Missouri' in row[0]:
                sname = 'Missouri'
            elif row[0].strip() in STATES:
                sname = row[0].strip()
            else:
                continue
        
            # create a dictionary with crop name as key and value as another dictionary
            # whose key is state name and value is :
            
            #                 {"Max_Year" : year,
            #                 "Max_val" : val,
            #                 "Min_Year" : year,
            #                 "Min_val" : val}
                    
            
            if crop not in mydict:
                mydict[crop] = {}
            
            year = int(row[4])
            val = int(row[6])
            stat = {"Max_Year" : year,
                "Max_val" : val,
                "Min_Year" : year,
                "Min_val" : val}
                
            if sname not in mydict[crop].keys():
                mydict[crop][sname] = stat
                
            # find min ,max and their respective years.
            else:
                if val > mydict[crop][sname]["Max_val"]:
                    mydict[crop][sname]["Max_val"] = val
                    mydict[crop][sname]["Max_Year"] = year
                elif val < mydict[crop][sname]["Min_val"]:
                    mydict[crop][sname]["Min_val"] = val
                    mydict[crop][sname]["Min_Year"] = year
                elif val == mydict[crop][sname]["Max_val"]:
                    mydict[crop][sname]["Max_Year"] = min (year,mydict[crop][sname]["Max_Year"])
                elif val == mydict[crop][sname]["Min_val"]:
                    mydict[crop][sname]["Min_Year"] = min (year,mydict[crop][sname]["Min_Year"])
    return mydict
```

```python
def sort_dictionary(d): # sort the dictionary key
    return sorted(d)
```

```python
def main():
    try: # to handle exception
        mydict = read_file(open_file())

    # print the output
        for crop in sort_dictionary(mydict):
            print("Crop:",crop)
            print("{:<20s}{:<8s}{:<6s}{:<8s}{:<6s}".format("State", "Max Yr", "Max", "Min Yr", "Min"))
            for state in sort_dictionary(mydict[crop]):
                print("{:<20s}{:<8s}{:<6d}{:<8s}{:<6d}".format(state, str(mydict[crop][state]["Max_Year"]),
                                                                mydict[crop][state]["Max_val"],
                                                                str(mydict[crop][state]["Min_Year"]),
                                                                mydict[crop][state]["Min_val"]))
            print()
    except:
        print('Unable to read the file.')

if __name__ == "__main__":
    main()
```
