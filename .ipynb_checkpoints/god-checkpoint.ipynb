```python
# import library
import matplotlib.pyplot as plt  
import numpy as np
```

```python
# create a function that will display the image 
# after applying the given filter
def apply_filter(image,K):
    padding=1
    # add the zero padding to the image
    padded_image = np.zeros((image.shape[0]+padding*2, image.shape[1]+padding*2, 3))
    padded_image[int(padding):int(-1 * padding), int(padding):int(-1 * padding),:] = image
    
    # for each channel apply the filter separately
    for z in range(padded_image.shape[2]): # channel
        minVal = padded_image[1:-1,1:-1,z].min() # min pixel value of a channel
        maxVal = padded_image[1:-1,1:-1,z].max()  # max pixel value of a channel
        for i in range(0,padded_image.shape[0]-2*padding): # row wise looping
            for j in range(0,padded_image.shape[1]-2*padding): # column wise looping
                x = np.sum(padded_image[i:i+3,j:j+3,z]*K)
                # handle out of range pixel,min-max normalization
                x = abs(x)%255
                x = ((x - minVal) / (maxVal - minVal)) * (255 - 0) + 0
                padded_image[i+1,j+1,z] = x # update the individual pixel
    # as,pixel value is float,convert image pixel into integer using uint8 
    plt.imshow(np.uint8(padded_image[1:-1,1:-1,:])) 
    plt.show()
```

```python
kernel1 = np.array([[0,0,-1],[0,0,0],[1,0,0]])
kernel2 = np.array([[0,0.1,0],[0.2,0.4,0.2],[0,0.1,0]])
kernel3 = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
```

```python
image=plt.imread('dog.JPG')
plt.imshow(image)
plt.show()
```

```python
apply_filter(image,kernel1) # filter 1
```

```python
apply_filter(image,kernel2) # filter 2
```

```python
apply_filter(image,kernel3) # filter 3
```
