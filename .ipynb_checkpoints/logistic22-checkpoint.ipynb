# Classification Problem

```python
# import the library
import pandas as pd

# download the dataset from 
# https://www.kaggle.com/uciml/breast-cancer-wisconsin-data/code

# read the dataset
dataset = pd.read_csv('data.csv')
dataset.head() # display few rows.
```

# data cleaning

```python
# drop the unwanted attributes.
dataset.drop(['id','Unnamed: 32'], inplace=True, axis=1)
dataset.head() # display few rows.
```

```python
# get the unique values for 'diagnosis' column
dataset['diagnosis'].unique()
```

```python
# encode value of 'diagnosis' column from M and B to 1 and 0 respectively
dataset['diagnosis'] = dataset['diagnosis'].apply(lambda x: 1 if x=='M' else 0)

dataset.head() # display few rows.
```

```python
# define X and y

# all the columns except 1st one is the feature column
X = dataset.iloc[:,1:].values

# only the first attribute (diagnosis) is the Target column
y = dataset.iloc[:,0].values 
```

```python
X
```

```python
y
```

# data splitting

```python
# define training and test set
# 70% data is for training and remaining 30% is for test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
```

# data normalization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# prediction

```python
# Build Logistic Regression Model

from sklearn.linear_model import LogisticRegression
logit_clf = LogisticRegression(random_state = 42)
logit_clf.fit(X_train, y_train)
```

```python
# prediction on test case
y_pred = logit_clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

# confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
conf_matrix
```

```python
accuracy_score(y_test, y_pred)
```

```python
print('Accuracy: {:0.3}'.format(accuracy_score(y_test, y_pred)))
```

```python
# classification report
print(classification_report(y_test, y_pred))
```
