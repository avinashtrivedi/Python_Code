- bins denotes the number of bins the dataset can be divided into while plotting the histogram

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.visualization import hist
x = np.random.randn(500)
plt.hist(x, bins=15) # set 15 bin
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.visualization import hist
x = np.random.randn(500)
plt.hist(x, bins=range(-3, 4)) # set bin edge at location -3,-2,....2,3
plt.show()
```

 - Matplotlib also provide the auto tuning of bins

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.visualization import hist
x = np.random.randn(500)
plt.hist(x, bins='auto') # auto tuning
plt.show()
```

- Freedman-Diaconis,Sturges',Scott's,the Square-root rule these are the few rule of thumb proposed to get the best number of bins.
- Reference: https://docs.astropy.org/en/latest/visualization/histogram.html

```python
# the above bins tuning can be implemented using astropy library.
from astropy.visualization import hist
hist(x, bins='scott') # using scott rule
plt.show()
```

```python
from astropy.visualization import hist
hist(x, bins='freedman') # using Freedman-Diaconis rule
plt.show()
```

- The Freedman-Diaconis and Scott’s rule make the bins tuning suboptimal for complex dataset.
- Hence astropy implemented another two rules that uses fitness function to find the optimal binning.
 - Knuth’s rule
 - Bayesian Blocks

```python
from astropy.visualization import hist
hist(x, bins='knuth') # using Knuth’s rule
plt.show()
```

```python
from astropy.visualization import hist
hist(x, bins='blocks') # using Bayesian Blocks rule
plt.show()
```
