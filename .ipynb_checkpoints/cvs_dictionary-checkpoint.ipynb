```python
# import library
import pandas as pd
```

```python
# create a function to get a paramater as filename and return dictionary 
# as per question

def CSVToDict(filename):
    
    # read the file
    df = pd.read_csv(filename,na_values='none')
    
    # get necessary columns
    df = df[['title','author','language','rating','publisher','page_count','generes']]
    
    # drop NaN rows
    df.dropna(inplace=True)
    
    # get the preprocessed genre names
    df.generes = df.generes.apply(lambda x : x.split(',')[0].split('&amp')[0])
    
    # groupby on the basis of 'generes' column
    mydict = {k: x[['title','author','language','rating','publisher','page_count']].to_dict('records') for k,x in df.groupby('generes')}
    return mydict
```

```python
# call the function
CSVToDict('google_books_1299.csv')
```
