```python
# read the file
with open('RealEstateData.txt', 'r') as file:
    em_data = file.readlines() 
```

```python
# extract the price column and store in prices list
prices = [i.strip().split()[0] for i in em_data if i.strip()!='']
```

```python
prices
```

```python
# as the price is in string form,convert it to int 
print('Before conversion: ',type(prices[0]))
prices = [int(price) for price in prices]
print('After conversion: ',type(prices[0]))
```

```python
# now all the price is in integer form
prices
```

```python
# mean function
def user_mean(prices):
    total =0
    for i in prices:
        total = total + i
    # total sum divided by length
    return total/len(prices)
```

```python
user_mean(prices)
```

```python
# median function
def user_median(prices):
    # length of prices list
    n = len(prices)
    # sort the prices in ascending order.
    prices.sort()
    # if number of element is even
    # then median is average of n/2 and ((n/2)+1)th element.
    # else element at index n/2
    if n%2==0:
        x = prices[n/2]
        y = prices[(n/2)+1]
        return (x+y)/2
    else:
        return prices[n/2]
```

```python
# mode function
# its is the most occurring element i.e element with highest count
def user_mode(prices):
    # to keep track of element with max count
    price_counter= 0  # initial count of element
    actual_price = 0 # initial price whose count is maximum
    # find unique element in list
    unique_prices = set(prices)
    
    # for each unique element in prices,compute its number of occurrence
    for i in unique_prices:
        # count the occurrence of each element.
        count = prices.count(i)
        # keep track of element with max count
        if count>price_counter:
            price_counter = count
            actual_price = i
    return actual_price
```

```python
user_mode(prices)
```
