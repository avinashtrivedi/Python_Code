```python
# import library
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
# save that excel file as csv in your current working directory
# and then read that csv file here
df = pd.read_csv('property_data.csv')
```

```python
df.columns # list of columns
```

```python
# select columns for Model creation and price prediction
# remaining column are of no use for price prediction like cid,zip code  etc..

list_of_features_and_target = ['price','room_bed', 'room_bath', 'living_measure',
                    'lot_measure', 'ceil', 'coast', 'sight', 'condition',
                     'quality','ceil_measure', 'basement',
                    'living_measure15', 'lot_measure15', 'furnished','total_area']
```

```python
# get the filtered dataframe
df = df[list_of_features_and_target]

# drop the NaN rows
df.dropna(inplace=True)
```

```python
df.head() # display few rows
```

```python
column = list(df)

# ignore rows containing '$'
for i in column:
    df = df[df[i]!='$']
```

```python
# get X and y
# X as a feature and y as a target
X = df.iloc[:,1:].values
y = df.iloc[:,0].values
```

```python
X
```

```python
y
```

```python
# Split the dataset into training and test set
# 70% for training and 30% for testing.

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Build Decision Tree regression Model 
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X_train,y_train) # fit the model on training data

# Prediction on Test data
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
# create a new dataframe containing
# True price and predicted price
df_new = pd.DataFrame({'true_price':y_test,'predicted_price':y_pred})
```
