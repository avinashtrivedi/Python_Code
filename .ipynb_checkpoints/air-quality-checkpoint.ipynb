```python
import pandas as pd # import library
```

```python
# source: https://www.kaggle.com/anveshparashar/airqualityuci
df = pd.read_csv('AirQualityUCI.csv')
```

```python
df.head() # display few rows
```

```python
# creating a lag by 3
# you can change this value according to your need
# Create two column T_lag and RH_lag

df['T_lag'] = df['T'].shift(-3)
df['RH_lag'] = df['RH'].shift(-3)
```

```python
df.head() # display few rows
```

```python
# Create two new column T_lag_rolling and RH_lag_rolling containing
# 10 hour rolling average of T_lag and RH_lag column respectively.

# 10 hour rolling average means rolling average of previous 10 records

df['T_lag_rolling'] = df['T_lag'].rolling(10).mean() # 10 hour rolling average for T_lag column
df['RH_lag_rolling'] = df['RH_lag'].rolling(10).mean()  # 10 hour rolling average RH_lag column

df['T_lag_rolling'] = round(df['T_lag_rolling'] ,2) # round to 2 decimal places
df['RH_lag_rolling'] = round(df['RH_lag_rolling'],2) # round to 2 decimal places
```

```python
df.head(20) # display top 20 rows
```

```python
df = df.dropna() # drop the rows with NaN value
```

```python
df.head(10) # display the top 10 rows
```

```python
df = df.reset_index(drop=True) # reset the index
df.head() # display few rows
```

```python
# exporting to csv file.
# this file will be saved in current working directory,i.e. where you have saved your python program
df.to_csv('Output.csv') 
```
