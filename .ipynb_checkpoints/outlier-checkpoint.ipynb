# Q.c

```python
# reset the index, so that it start from 0
df.reset_index(drop=True,inplace=True)
```

```python
df # display the data 
```

```python
# print row and column of Missing Values.

for i in range(19):
    for j in range(5):
        if df.iloc[i,j] == '.':
            print('Missing Value at (row,col): ({}, {}) '.format(i,j))
```

# Q.d

```python
# replace Missing Values with NaN for simplicity

df.replace('.',float('NaN'),inplace=True)
```

```python
df #display the data 
```

```python
# get the gender with high frequency
gender_percent = df['Gender'].value_counts()
gender_high_freq = gender_percent.index[0]
```

```python
gender_high_freq
```

```python
# fill the NaN of Gender column with gender_high_freq
df['Gender'] = df['Gender'].fillna(gender_high_freq)
```

```python
df
```

```python
df.dtypes
```

```python
# change the datatype of Age,Height and Weight columns to float
df['Age'] = df['Age'].astype('float')
df['Height'] = df['Height'].astype('float')
df['Weight'] = df['Weight'].astype('float')
```

```python
# fill NaN here with its median
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df
```

```python
# fill NaN of Height and Weight column with its mean
df['Height'] = df['Height'].fillna(df['Height'].mean())
df['Weight'] = df['Weight'].fillna(df['Weight'].mean())
```

```python
df
```

# Q.e

```python
# remove outlier from Age column
QLow = df["Age"].quantile(0.01)
QHi  = df["Age"].quantile(0.99)
df = df[(df["Age"] < QHi) & (df["Age"] > QLow)]
```

```python
df
```

```python
# remove outlier from Height column

QLow = df["Height"].quantile(0.01)
QHi  = df["Height"].quantile(0.99)

df = df[(df["Height"] < QHi) & (df["Height"] > QLow)]
```

```python
df
```

```python
# remove outlier from Weight column

QLow = df["Weight"].quantile(0.01)
QHi  = df["Weight"].quantile(0.99)

df = df[(df["Weight"] < QHi) & (df["Weight"] > QLow)]
```

```python
df
```

```python
# reset the index
df.reset_index(drop=True,inplace=True) 

df # final dataframe
```
