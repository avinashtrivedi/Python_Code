```python
while True:
    
    print('Enter 4 non-negative integers in range 0-255\n')
    
    # getting the inputs as string
    while True:

        a = input('1st Non-negative integer: ')
        if a.isdigit():
            if 0<=int(a)<=255:
                break
            else:
                print('Enter integer in the range 0-255, Try again.')
        else:
            print('Please enter Non-negative integer only. Try again.')
        
    while True:

        b = input('2nd Non-negative integer: ')
        if b.isdigit():
            if 0<=int(b)<=255:
                break
            else:
                print('Enter integer in the range 0-255, Try again.')
        else:
            print('Please enter Non-negative integer only. Try again.')
    
    while True:

        c = input('3rd Non-negative integer: ')
        if c.isdigit():
            if 0<=int(c)<=255:
                break
            else:
                print('Enter integer in the range 0-255, Try again.')
        else:
            print('Please enter Non-negative integer only. Try again.')
    
    while True:

        d = input('4th Non-negative integer: ')
        if d.isdigit():
            if 0<=int(d)<=255:
                break
            else:
                print('Enter integer in the range 0-255, Try again.')
        else:
            print('Please enter Non-negative integer only. Try again.')
        
    print('\nFirst Non-negative integer',a)
    print('Second Non-negative integer',b)
    print('Third Non-negative integer',c)
    print('Fourth Non-negative integer',d)
    
    # form the ipv4 address
    ipv4 = a + '.' + b + '.' + c + '.' + d
    
    print('\nIPv4 address:',ipv4)
    
    while True:
        choice = input('\nDo you want to continue ? (Y)es or (N)o: ')
        if choice=='N':
            flag = False
            break
        elif choice!='Y':
            print('Wrong input, Please try again.')
        else:
            flag = True
            break
    print()
    if flag==False:
        break
```
