```python
def gets(s):
    try:
        if s[2] in var:
            var[s[0]] = var[s[2]]
        else:
            var[s[0]] = int(s[2])
    except:
        print('Syntax error.')

def adds(s):
    try:
        if s[0] in var and s[2] in var:
            var[s[0]]= var[s[0]] + var[s[2]] 
        elif s[0] in var:
            var[s[0]]= var[s[0]] + int(s[2])
        else:
            print(s[0],'is undefined')
    except:
        print('Syntax error.')

def inputs(s):
    try:
        var[s[1]] = int(input("Enter a value for {}: ".format(s[1])))
    except:
        print('Syntax error.')
        
def prints(s):
    try:
        if s[1] in var:
            print(s[1],'equals',var[s[1]])
        elif int(s[1]):
            print(s[1])
    except:
        print(s[1],'is undefined')
```

```python
print('Welcome to the Adder REPL')
quit=True # to handle the while loop
var={} # create a empty dictionary to store the variables defined during runtime. 
# store variable name as a key in dictionary and variable value as value of corresponding key. 
while quit:
    try:
        s = input()
        s = s.split(' ') # create a list of string eg.  a gets 1  become  ['a','gets','1']
        # handle cases like -->  a gets 1 , c gets a ,23 gets z
        if len(s)==3 and s[1]=='gets':
            try:
                if s[2] in var:
                    var[s[0]] = var[s[2]]
                else:
                    var[s[0]] = int(s[2])
            except:
                print('Syntax error.')
        # handle cases like -->  c adds b
        elif len(s)==3 and s[1]=='adds':
            try:
                if s[0] in var and s[2] in var:
                    var[s[0]]= var[s[0]] + var[s[2]] 
                elif s[0] in var:
                    var[s[0]]= var[s[0]] + int(s[2])
                else:
                    print(s[0],'is undefined')
            except:
                print('Syntax error.')
        # handle cases like --> input b
        elif len(s)==2 and s[0]=='input':
            try:
                var[s[1]] = int(input("Enter a value for {}: ".format(s[1])))
            except:
                print('Syntax error.')

        # handle cases like --> print a,print z,print 32
        elif len(s)==2 and s[0]=='print':
            try:
                if s[1] in var:
                    print(s[1],'equals',var[s[1]])
                elif int(s[1]):
                    print(s[1])
            except:
                print(s[1],'is undefined')
        # handle quit
        elif len(s)==1 and s[0]=='quit':
            quit=False
            print('Goodbye.')
        # handle any cases not handled earlier.
        else:
            print('Syntax error.')
    except:
         print('Syntax error.')
```
