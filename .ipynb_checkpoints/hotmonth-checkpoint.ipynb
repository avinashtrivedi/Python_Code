```python
# load toolkits

import pandas as pd

import numpy as np

#create daily data to start

df = pd.DataFrame({

    'date': pd.date_range("2020-01-01", periods=365, freq='D'),

    'temperature': np.random.randint(0, 43, 365) + np.random.random(365).round(1)

})

# set date to index

df.set_index('date', inplace=True)

# convert to weekly
df = df.resample('W').mean()
```

```python
# reset the index
df.reset_index(inplace=True)

# display few rows
df.head()
```

```python
# create Month column
df['Month'] = df['date'].apply(lambda x: x.month)

# create week_number column
df['week_number'] = df.groupby(['Month']).cumcount()+1
```

```python
df.head()
```

```python
# get the hottest week of a month

df = df.loc[df.groupby('Month')['temperature'].idxmax()]

# set Month column as index
df.set_index('Month', inplace=True)
df
```
