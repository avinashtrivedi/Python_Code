```python
df = pd.read_csv('geoplaces2.csv')
```

```python
t=df['smoking_area'].value_counts()
t
```

```python
list(t)
```

```python
list(t.index)
```

```python
import pandas as pd
from matplotlib import pyplot as plt
 
# Figure Size
fig = plt.figure(figsize =(10, 7))
plt.title('Bar plot')
# Horizontal Bar Plot
plt.bar(list(t.index), list(t),label='smoking_area')
plt.xlabel('smoking area')
plt.ylabel('Count')
plt.legend()

    
for index, data in enumerate(list(t)):
    plt.text(x=index , y =data+1 , s=f"{data}",fontdict=dict(fontsize=15))
# Show Plot
plt.show()
```

```python
df = pd.read_csv('userprofile.csv')
```

```python
g = df['drink_level'].value_counts()
```

```python
x = list(g.index)
x
```

```python
y = list(g)
y
```

```python
plt.stackplot(x,y)
  
# Days
plt.xlabel('Drinking levels')
  
# No of hours
plt.ylabel('Count')
  
# Title of Graph
plt.title('Stackplot')
plt.legend(labels='Drinking levels')

for index, data in enumerate(y):
    plt.text(x=index , y =data+1 , s=f"{data}",fontdict=dict(fontsize=10))
# Displaying Graph
plt.show()
```
