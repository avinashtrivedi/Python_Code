```python
def combine(f,op,n):
    result = f(0)
    for i in range(n):
        result = op(result,f(i))
    return result
```

```python
def smiley_sum(t):
    def f(x):
        return x**2
    def op(x , y):
        return x + (y if y==1 else 2*y)
    n = t+1
    return combine(f,op,n)
```

```python
smiley_sum(1)
```

```python
smiley_sum(2)
```

```python
smiley_sum(3)
```

```python
smiley_sum(4)
```
