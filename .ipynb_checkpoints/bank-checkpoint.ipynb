```python
try: # to handle exception
    account={} # Dictionary to hold the account
    output=[] # store the output
    while True:
        x = input()
        if x=='':
            break
        x = x.split(' ') # split the input to get a list
        if len(x)==2:
            if x[0]=='BALANCE':
                if x[1] in account:
                    output.append(account[x[1]])
                else:
                    output.append('ERROR')
            elif x[0]=='INCOME':
                for i in account:
                    if account[i]>=0:
                        account[i] = account[i] + int(account[i]*int(x[1])/100)
            else:
                print('wrong input')
                break
        elif len(x)==3:
            if x[0]=='DEPOSIT':
                if x[1] in account:
                    account[x[1]] = account[x[1]] + int(x[2])
                else:
                    account[x[1]] = 0
                    account[x[1]] = account[x[1]] + int(x[2])
            elif x[0]=='WITHDRAW':
                if x[1] in account:
                    account[x[1]] = account[x[1]] - int(x[2])
                else:
                    account[x[1]] = 0
            else:
                print('wrong input')
                break
        elif len(x)==4:
            if x[0]=='TRANSFER':
                if x[1] in account and x[2] in account:
                    account[x[1]] = account[x[1]] - int(x[3])
                    account[x[2]] = account[x[2]] + int(x[3])
                elif x[1] in account and x[2] not in account:
                    account[x[2]] = 0
                    account[x[1]] = account[x[1]] - int(x[3])
                    account[x[2]] = account[x[2]] + int(x[3])
                elif x[1] not in account and x[2] in account:
                    account[x[1]] = 0
                    account[x[1]] = account[x[1]] - int(x[3])
                    account[x[2]] = account[x[2]] + int(x[3])
                elif x[1] not in account and x[2] not in account:
                    account[x[1]] = 0
                    account[x[2]] = 0
                    account[x[1]] = account[x[1]] - int(x[3])
                    account[x[2]] = account[x[2]] + int(x[3])
            else:
                print('wrong input')
                break
        else:
            print('wrong input')
            break
    # if there is some output
    if len(output)>=1:
        print('Program output:')
        for i in output:
            print(i)
    else:
        print('Try again')
except:
    print('wrong input')
```
