# Number of trees 
- Method :1

```python
# How many trees in each arrondissement (ARROND_NOM)
# To know the number of trees in each arrondissement (Borough):
# count the number of rows with same arrondissement
# example: count the number of rows with 'Verdun' as arrondissement ,
# its count becomes the number of trees in 'Verdun'
# similarly ,count the number of rows for another arrondissement.

# value_counts: return the count of unique rows,in other words
# count of each unique arrondissement in the dataset
no_tree = data['ARROND_NOM'].value_counts()
no_tree
```

```python
# print with proper formatting
print("{:<70} {:<10}".format('arrondissement','No of Tree'))
print(90*"*") # print * 90 times
# print number of trees in each arrondissement
for index, val in no_tree.iteritems():
    print("{:<70} {:<10}".format(index,val))
```

# Number of Trees 
- Method :2

```python
# group the dataset using ARROND_NOM i.e grouping together each unique value of ARROND_NOM column and find the count.
data.groupby(by=['ARROND_NOM']).agg(['count'])
```

```python
# then take ARROND column and display its count
data.groupby(by=['ARROND_NOM'])['ARROND'].agg(['count'])
# Here, count column represents the number of trees in each arrondissement
```
