```python
# import the library
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# load the dataset
iris = datasets.load_iris()
data = iris['data']
label = iris['target']
```

```python
wcss = [] # store "within cluster sum of squares" (WCSS)

# Apply K-Means for K = 2 to K = 5
for i in range(2, 6):
    kmeans = KMeans(n_clusters = i,random_state = 42)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
    
# draw the elbow plot
# to know what value of K is best

plt.plot(range(2, 6), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # within cluster sum of squares
plt.show()
```

- At K=5, within cluster sum of squares (WCSS) is lowest, hence this is the optimum amount of clusters

```python
# now build K-means Model on K=5
kmeans = KMeans(n_clusters = 5,random_state = 42)
y_pred = kmeans.fit_predict(data)
```

```python
# visualize the clusters
plt.scatter(data[y_pred == 0, 0], data[y_pred == 0, 1],  c = 'yellow', label = 'Iris-setosa')
plt.scatter(data[y_pred == 1, 0], data[y_pred == 1, 1],  c = 'blue', label = 'Iris-versicolour')
plt.scatter(data[y_pred == 2, 0], data[y_pred == 2, 1],  c = 'green', label = 'Iris-virginica')

# Plot the centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1],marker='*', s = 100, c = 'red', label = 'Centroids')
plt.legend()
plt.show()
```
