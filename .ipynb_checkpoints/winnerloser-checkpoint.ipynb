```python
# import the library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# read the dataset
df = pd.read_csv('RegularSeasonCompactResults.csv')
```

# Q.1

```python
# create a total_score column
df['total_score'] = df['Wscore'] + df['Lscore']
```

# Q.3

```python
# average total score for each year
avg_total_score = df.groupby(by=['Season'])[['total_score']].mean()
avg_total_score
```

# Q.4

```python
plt.plot(avg_total_score,color='r')
plt.xlabel('Year')
plt.ylabel('Average total score')
plt.show()
```

# Q.5

```python
# create new dataframe
df_new = df[['Season','Wscore','Lscore']]

# create 'Margin' column 
df_new['Margin'] = df['Wscore'] - df['Lscore']

# create another column 'Margin_30_or_more' with 1 where margin >=30 else 0
df_new['Margin_30_or_more'] = np.where(df_new['Margin']>=30,1,0)

# compute percentage of blowout 
df_blowout_percent = df_new.groupby(['Season'])['Margin_30_or_more'].apply(lambda x: sum(x)*100/x.count())
```

```python
df_blowout_percent
```

```python
# get year with highest percentage of blowout games
df_blowout_percent.sort_values(ascending=False).head(1)
```

```python
# you can also compute using this code

indx = df_blowout_percent.argmax()
percent = df_blowout_percent.iloc[indx]
year = df_blowout_percent.index[indx]

print('Year: {} has highest percentage({:.2f}%) of blowout games'.format(year,percent))
```

# Q.6

# a

```python
# get data for winner and loser
df_winner = df[['Season','Wteam','Wscore']]
df_loser = df[['Season','Lteam','Lscore']]
```

```python
# make the same column name
df_winner.columns = ['Season', 'TeamID', 'Score']
df_loser.columns = ['Season', 'TeamID', 'Score']
```

```python
# append both the dataframe (teams i.e. winner and loser)
df_all_team = df_winner.append(df_loser,ignore_index=True)
```

```python
df_all_team
```

```python
# For each team and season combination, average number of points scored by the team 
df_all_team_avg_points = df_all_team.groupby(['TeamID','Season']).mean()
df_all_team_avg_points
```

```python
# reset the index to make it single indexed dataframe
df_all_team_avg_points =  df_all_team_avg_points.reset_index()
df_all_team_avg_points
```

# b

```python
# get the data for team 1461
df_team_1461 = df_all_team_avg_points[df_all_team_avg_points['TeamID']==1461][['Season','Score']]
```

```python
df_team_1461
```

```python
# set the 'Season' column as index
df_team_1461.set_index('Season', inplace=True)
df_team_1461
```

```python
# plot the graph
plt.plot(df_team_1461,'r')
plt.xlabel('Year')
plt.ylabel('Average point')
plt.show()
```
