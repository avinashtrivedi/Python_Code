```python
while True:
    
    four_num = input('Enter four numbers separated by Comma: ')
    
    # split the input string
    four_num = four_num.split(',')
    
    # check if number of input is 4
    if len(four_num)==4:
        
        # try-except block to handle exception.
        try:
            # convert string numbers to float
            four_num = [float(i) for i in four_num]

            first_num = four_num[0]
            second_num = four_num[1]
            third_num = four_num[2]
            fourth_num = four_num[3]

            # compute result
            result = ((first_num*second_num) + third_num)/fourth_num
            break

        except ValueError:
            print('Invalid Input,Please enter four valid numbers separated by Comma.')
            continue

        except ZeroDivisionError:
            print('Fourth number is Zero,try again.')
    else:
        print('Invalid Input,Please enter four valid numbers separated by Comma.')

print()
print('(({}*{}) + {})/{} = {:0.3}'.format(first_num,second_num,third_num,fourth_num,result))
```

```python
import random

import math

'''find prime'''

def prime(num):

  if num <= 1:

    return False

  for i in range(2, int(math.sqrt(num)) + 1):

    if (num % i) == 0:

      return False

    else:

      return True



'''find random prime'''

def rollUntil3ConsPrimes():

  rolls = [0, 0, 0]

  while True:

    rolls[0] = random.randint(1, 6)

    rolls[1] = random.randint(1, 6)

    rolls[2] = random.randint(1, 6)

    if prime(rolls[0]) and prime(rolls[1]) and prime(rolls[2]):

      break

  return rolls



'''Finding ending score'''

def rollUnitll():

  score=100

  Rolls= []

  count=0

  while True:

    dicenum= rollUntil3ConsPrimes()

    Rolls.append[dicenum]

    

     

    if rollUntil3ConsPrimes():

      count +=1

    if count == 3:

      break

     

    else:

      score -=5

  return Rolls,score



print('Sample runs:',dicenum)

print('Rolls made:',Rolls)

print('score=',score)
```
