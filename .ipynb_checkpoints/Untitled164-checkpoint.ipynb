```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

###Speech Files###
sessions = np.arange(25, 76)
data=[]


for session in sessions:
#     directory = "./TXT/Session "+str(session)+" - "+str(1945+session)
    directory = r"E:\New folder\TXT\Session "+str(session)+" - "+str(1945+session)
    for filename in os.listdir(directory):
        f = open(os.path.join(directory, filename), encoding='utf8')
        if filename[0]==".": #ignore hidden files
            continue
        splt = filename.split("_")
        data.append([session, 1945+session, splt[0], f.read()])
        
df_speech = pd.DataFrame(data, columns=['Session','Year','ISO-alpha3 Code','Speech'])


df_speech.tail(10)
```

```python
###Methodology file### 
df_codes = pd.read_csv("UNSD — Methodology.csv", header = None, sep = '\n')
df_codes = df_codes[0].str.split(',', expand =True)
df_codes.tail()


new_header = df_codes.iloc[0]
df_codes = df_codes[1:]
df_codes.columns = new_header
df_codes.tail()



####(3) Word Count by Country between 2005-2020####
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk import FreqDist


def preprocess(words):
    words = [word.lower() for word in words]
    words = [word for word in words if word.isalpha()]
    sw = stopwords.words("english")
    no_sw = []
    for w in words:
        if (w not in sw):
            no_sw.append(w)
    return no_sw


ISO_alpha3_Code = list(df_speech['ISO-alpha3 Code'].unique())


Total_Word_Count = []
for each in ISO_alpha3_Code:
    t = df_speech.loc[(df_speech['Year'] > 2005) & (df_speech['Year'] < 2020) & (df_speech['ISO-alpha3 Code'] == each)]['Speech']
    if(t.empty):
         Total_Word_Count.append(0)
    else:
        text=t.values[0]
        words = word_tokenize(text)
        words = preprocess(words)
        freq = FreqDist(words)


        words1 = words.count("freedom")
        words2 = words.count("GDP")
        words3 = words.count("support")
        words4 = words.count("health")
        words5 = words.count("generosity")
        total = words1+words2+words3+words4+words5
        Total_Word_Count.append(total)


new=pd.DataFrame(zip(ISO_alpha3_Code,Total_Word_Count),columns=['ISO-alpha3 Code',"Total Word Count"])
df_codes=pd.merge(df_codes, new, on='ISO-alpha3 Code',how='left')
df_codes.tail()
```
