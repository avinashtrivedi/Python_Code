```python
# import library
import subprocess
import socket
import grp
import re
import pwd

# create a txt file object with write mode
file_object = open("details.txt", "w")

# function to get machine name
def MachineName():
    machine_name = socket.gethostname()
    file_object.write(f"Machine Name is: {machine_name}\n")

# function to get user and its corresponding group
def UserGroup():
    file_object.write(f"{'User'.ljust(31)} {'Group'.ljust(32)}\n")
    for var in pwd.getpwall():        
        file_object.write(f"{var[0].ljust(31)},{grp.getgrgid(var[3])[0].ljust(32)}\n")
       
 # function to get processor information
def processor_info():
    
    # linux command to get the cpu information
    p_command = "cat /proc/cpuinfo" 
    # run this command in python using subprocess
    p_info = subprocess.check_output(p_command, shell=True).strip()
    p_info = p_info.decode('UTF-8')
    p_info = p_info.split("\n")
    file_object.write("\n") 
    # writing a processor information into file
    for i in p_info:
        if "processor" in i:
            file_object.write("\n") 
            file_object.write(f'Processor:{re.sub( ".*processor.*:", "", i,1)}\n')      # processor details
        if "vendor_id" in i:
            file_object.write(f'Vendor id:{re.sub( ".*vendor_id.*:", "", i,1)}\n')      # vendor id details
        if "model" in i:
            file_object.write(f'Model:{re.sub( ".*model.*:", "", i,1)}\n')              # model details
        if "model name" in i:
            file_object.write(f'Model name:{re.sub( ".*model name.*:", "", i,1)}\n')    # model name details
        if "cache size" in i:
            file_object.write(f'cache size:{re.sub( ".*model name.*:", "", i,1)}\n')    # cache details
    
def services():
    # linux command to get the service and its status
    s_command = "systemctl list-unit-files"  
    # run the command
    s_info = subprocess.check_output(s_command, shell=True).strip()     
    s_info = s_info.decode('UTF-8')
    file_object.write("\n")
    # Writing the data to file
    file_object.write(s_info) 
    
def main():
    MachineName() # machine name
    
    ###### for better formatting ################
    file_object.write("user and associated group: \n")
    file_object.write("\n") 
    file_object.write(60*"-")  # writing '-' 60 times for better formatting
    file_object.write("\n") 
    ############################################
    
    UserGroup() # user and associated group
    
    file_object.write(60*"-") 
    file_object.write("\n")
    
    processor_info() # processor information
    
    file_object.write(60*"-") 
    file_object.write("\n")
    
    services() # services and its status
    
    file_object.write("\n")
    file_object.write(60*"-") 
    file_object.write("\n")
    # close the file
    file_object.close() 
    print("file details.txt is created and saved in current directory")
    
# call the main function
if __name__ == '__main__':
    main()
```
