```python
def main():
    pension_rate = .12 # pension rate of 12%
    flag=True # flag to control outer while loop
    emp_details=[] # store individual employee details
    while flag:
        name = input("Enter employee's name: ")
        pay_rate = float(input("Pay Rate: $"))
        hour_worked = float(input("Hours worked: "))
        tax_rate = float(input("Tax deduction rate: "))
        # store employee details as tuple of (name,pay rate,hour worked,tax rate)
        emp_details.append((name,pay_rate,hour_worked,tax_rate))
        print("")
        
        # enter 'y' to continue or 'n' to exit the program
        
        # Also control wrong input, i.e other than y/n
        # if user entered anything other than y or n 
        # he will be prompted to "wrong input,Please try again.."
        while True:
            command = input("Add the details of another employee y/n? ")
            if command=='n':
                flag=False
                break
            elif command=='y':
                break
            elif command not in ('y','n'):
                print("wrong input,Please try again..")
    print(100*"-") # print '-' 100 times for clear formatting
    # {:<15} : used for assigning 15 spaces for printing.
    print("{:<20} {:<20} {:<20} {:<20} {:<20}".format('Name','Gross Pay','Tax','Pension','Net Pay'))
    print(100*"-")
    
    # loop through each item of emp_details
    # it is a list of tuples, where first element of tuple is name ,second element is pay rate
    # third element is hour worked and fourth element is tax rate.
    for i in emp_details:
        Name = i[0] # name
        gross_pay = i[1]*i[2] # pay_rate * hour_worked
        tax = (gross_pay*i[3])/100 # gross_pay * tax_rate
        pension = gross_pay*pension_rate 
        net_pay = gross_pay - tax - pension
        
        # output formatting
        net_pay_out = "$" + format(net_pay, ',.2f')
        gross_pay_out = "$" + format(gross_pay, ',.2f')
        tax_out = "$" + format(tax, ',.2f')
        pension_out = "$" + format(pension, ',.2f')
        
        print("{:<20} {:<20} {:<20} {:<20} {:<20}".format(Name,gross_pay_out,tax_out,pension_out,net_pay_out))
    print(100*"-")
main()
```
