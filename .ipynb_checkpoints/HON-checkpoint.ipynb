```python
# import the library
import pandas as pd
import numpy as np

# read the dataset downloaded from yahoo finance
df = pd.read_csv('HON.csv')

# display top first 10 rows
df.head(10)
```

```python
# compute the return
df['Return'] = df['Adj Close'].pct_change()
df.fillna(0,inplace=True)

# display top first 10 rows
df.head(10)
```

```python
# Create 'True Label' column
df['True Label'] = np.where(df['Return']>=0,'+','-')

# display top first 10 rows
df.head(10)
```
