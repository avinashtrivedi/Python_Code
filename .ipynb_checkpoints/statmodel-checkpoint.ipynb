```python
import pandas as pd
```

```python
df = pd.read_csv('listings.csv')
```

```python
df.columns
```

```python
df = df[['neighbourhood_cleansed', 'property_type', 'room_type', 'accommodates','bedrooms', 'beds', 'minimum_nights', 'maximum_nights','availability_365', 'number_of_reviews','review_scores_location']]
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
X = df.iloc[:, 3:-1].values
y = df.iloc[:, -1].values
X.shape
```

```python
import statsmodels.api as sm
import numpy as np
X = np.append(arr=np.ones((15019, 1)).astype(int), values=X, axis=1)
X_opt = X[:, [0, 1, 2, 3, 4, 5]]
regressor_OLS = sm.OLS(endog=y, exog=X_opt).fit()
regressor_OLS.summary()
```
