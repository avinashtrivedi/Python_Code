```python
# import the library
import pandas as pd
import matplotlib.pyplot as plt

# This website is down time and again
# df = pd.read_csv('https://philchodrow.github.io/PIC16A/datasets/gapminder.csv')

# Hence Downloaded and using a local copy
df = pd.read_csv('gapminder.csv')
```

```python
# Create a figure and a set of subplots with 1 row and 5 columns
# figsize=(20,3) : set the figure size
# sharey=True : Y axis will be same for all the plot

fig, (ax1, ax2,ax3,ax4,ax5) = plt.subplots(1, 5,figsize=(20,3),sharey=True)

# dictionary to hold continent and Axes object
d={'Asia': ax1,
   'Africa': ax2,
   'Europe': ax3,
   'Americas': ax4,
   'Oceania': ax5
  }

# dictionary to hold continent and color
c={'Asia': 'blue',
   'Africa': 'green',
   'Europe': 'yellow',
   'Americas': 'orange',
   'Oceania': 'deeppink'
  }

# this lambda function will plot the graph on different subplots.
subplot = lambda x: d[list(x['continent'])[0]].plot(x['year'],x['lifeExp'],color=c[list(set(x['continent']))[0]])

# this lambda function will set the title for each subplot
title = lambda x: d[list(x['continent'])[0]].set_title(list(set(x['continent']))[0])

# this is the main lambda function which call the previous two lambda function.
f = lambda x : (subplot(x),title(x))

df.groupby(["country"]).apply(f)

# set y label
ax1.set_ylabel('Life Expectancy (Years)')

# show the plot                            
plt.show()
```
