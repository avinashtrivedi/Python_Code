```python
# import the library
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the data
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# target names
iris.target_names
```

```python
# feature names
iris.feature_names
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

# Distance metric : euclidean or L2 Norm

```python
# Fit the clasifier to the Training set
# Load the necessary libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

# Build KNN with k=5 (default value)
classifier = KNeighborsClassifier(n_neighbors=5,metric = 'euclidean')

# Fit the model
classifier.fit(X_train, y_train)

# Predict on the Test set
y_pred = classifier.predict(X_test)

acc = accuracy_score(y_test, y_pred)*100
print('Accuracy of model (using euclidean distance) is:',round(acc, 3),'%')

# list to store cv scores
cv_scores = []

# apply 10-fold cross validation
for k in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())

plt.plot(range(1,30), cv_scores)
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.show()

best_k =  cv_scores.index(max(cv_scores)) + 1
print('Maximum accuracy is achieved when k =',best_k)
```

```python
# Fit the clasifier to the Training set
# Load the necessary libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

# Build KNN with k=5 (default value)
classifier = KNeighborsClassifier(n_neighbors=5,metric = 'chebyshev')

# Fit the model
classifier.fit(X_train, y_train)

# Predict on the Test set
y_pred = classifier.predict(X_test)

acc = accuracy_score(y_test, y_pred)*100
print('Accuracy of model (using chebyshev distance) is:',round(acc, 3),'%')

# list to store cv scores
cv_scores = []

# apply 10-fold cross validation
for k in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())

plt.plot(range(1,30), cv_scores)
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.show()

best_k =  cv_scores.index(max(cv_scores)) + 1
print('Maximum accuracy is achieved when k =',best_k)
```
