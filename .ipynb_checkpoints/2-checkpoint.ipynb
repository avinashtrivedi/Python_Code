4) Now applying i//(i%5)  will return <i>ZeroDivisionError</i> Because here we are diving i by zero. Since (i%5) give us 1 2 3 4 0 1 2 3 4 0 .For example

```python
1//0 # Giving error ZeroDivisionError,Just to show you how it is working
```

5) To Handle this error we need to use try and except block to,except block will be executed when try block through some error. Here i have added print(i,'hello') to show when except block is getting called.Finally it is printing value of i//(i%5) for i between 1 to 10.Whenever ZeroDivisionError error occured control exit from try and except block executed.

```python
d={(0,0):4,(2,3):7,(7,1):2,(6,6):9,(1,0):6,(0,4):4}
for v in d:
    if v[0] >5 and v[1]>5: # Modified line
        print('Co-ordinate > 5 :  ',v)
```
