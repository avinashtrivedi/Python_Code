```python
# import the library
import pandas as pd
```

```python
# load the dataset
y_79_y20_frame = pd.read_csv("data_79_20.csv")
```

```python
# preprocessing steps
if 'Unnamed: 0' in y_79_y20_frame:
    
    # set year as index
    y_79_y20_frame.set_index('Unnamed: 0', inplace=True)
    
    # remove index name
    y_79_y20_frame.index.name=''
```

```python
# display few rows
y_79_y20_frame.head()
```

```python
# define extract_fig_1_frame
def extract_fig_1_frame(DataFrame):
    
    # get the mean
    mean = DataFrame.mean()
    
    # get the std
    std = DataFrame.std(ddof=1)
    
    # create a new dataframe
    df = pd.DataFrame([mean,2*std],index=['mean','two_s'])
    
    # return df
    return df
```

```python
# call the function.
extract_fig_1_frame(y_79_y20_frame)
```
