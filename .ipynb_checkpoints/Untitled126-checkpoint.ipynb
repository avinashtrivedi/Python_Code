```python
# Import the required library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
wine_file = "https://raw.githubusercontent.com/ksynnott/VM-DS-Data/master/winequality.csv"
df = pd.read_csv(wine_file, sep=';')

# set the figure size
plt.figure(figsize=(18,4))

# draw the plot
for i,col in enumerate(['pH','density','alcohol','sulphates']):
    
    # creating a subplot
    plt.subplot(1,4,i+1)
    
    # drawing the distplot
    ax = sns.distplot(df[col],color='b',kde=True)

    # set x and y label
    ax.set_xlabel(None)
    ax.set_ylabel(None)
    title_txt = 'residual sugar' if col=='density' else col
    
    # setting the title
    ax.set_title(title_txt)

    plt.tight_layout()
```
