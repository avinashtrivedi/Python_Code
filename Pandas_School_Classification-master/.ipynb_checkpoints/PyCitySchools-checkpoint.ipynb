```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data = "schools_complete.csv"
student_data = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data_df = pd.read_csv(school_data)
student_data_df = pd.read_csv(student_data)

# Combine the data into a single dataset
school_data_complete = student_data_df.merge(school_data_df, how="left", on=["school_name", "school_name"])
#school_data_complete.head()
```

```python
#set avgerage math
avg_math = round(school_data_complete["math_score"].mean(),2)
avg_math
```

```python
#set avgerage reading
avg_reading = round(school_data_complete["reading_score"].mean(),2)
avg_reading
```

```python
#create bins for pass fail groupings
bins =[0,69,100]
groupnames = ["Fail","Pass"]
```

```python
#pfm = pass/fail math
school_data_complete["pfm"]= pd.cut(school_data_complete["math_score"], bins, labels = groupnames)
#school_data_complete.head()
```

```python
#grouped by pass/fail for math pfm_df = pass/failed math
pfm_df = (school_data_complete.groupby(["pfm"])
                .apply(lambda x: (x["pfm"]== "Pass").count())
                .reset_index(name="count"))
#print(pfm_df)
```

```python
pass_math = pfm_df.loc [pfm_df["pfm"] == "Pass", :]
#print(pass_math)
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
#pctpm = % passing math
pctpm = (pass_math["count"])/student_data_df["student_name"].count()*100
#pctpm
```

```python
#pfr = pass/fail reading
school_data_complete["pfr"]= pd.cut(school_data_complete["reading_score"], bins, labels = groupnames)
#school_data_complete.head()
```

```python
#grouped by pass/fail for readin pfr_df = pass/failed reading
pfr_df = (school_data_complete.groupby(["pfr"])
                .apply(lambda x: (x["pfr"]== "Pass").count())
                .reset_index(name="count"))
#print(pfr_df)
```

```python
#count for students passing reading
pass_reading = pfr_df.loc [pfr_df["pfr"] == "Pass", :]
#print(pass_reading)
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
#pctpm = % passing reading
pctpr = (pass_reading["count"])/student_data_df["student_name"].count()*100
#pctpr
```

```python
#Create a dataframe to hold the above results


district_summary_df = pd.DataFrame({"Schools":[school_data_df["school_name"].count()]
                                    , "Student Count":[student_data_df["student_name"].count()]
                                    , "Total Budget":  ([school_data_df["budget"].sum()])#.map('${:,.2f}'.format)
                                    , "Avg Math Score":[round(school_data_complete["math_score"].mean(),2)]
                                    , "Avg Reading Score":[round(school_data_complete["reading_score"].mean(),2)]
                                    , "Passing Rate":[round((avg_math+avg_reading)/2, 2)]
                                    , "Percent Passing Math" : pctpm.map('{:.2f}%'.format)
                                    , "Percent Passing Reading" : pctpr.map('{:.2f}%'.format)
                                   })

district_summary_df
```

```python
school_data_complete= school_data_complete.rename(index = str, columns ={"type": "school_type"})
#school_data_complete.head()
```

```python
#group by school
school_summary = school_data_complete.groupby(["school_name"])
#school_summary.head(3)
```

```python
student_count = school_summary["student_name"].count()
#student_count
```

```python
total_passing_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['student_name'].count()
#total_passing_math
```

```python
percent_pass_math = percent_pass_math = (total_passing_math / student_count) * 100
#percent_pass_math
```

```python
total_passing_reading = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['student_name'].count()
#total_passing_reading
```

```python
percent_pass_reading  = (total_passing_reading / student_count) * 100
#percent_pass_reading
```

```python
school_summary_table = pd.DataFrame({ "School Type": school_summary["school_type"]
                                     ,"Total Students": school_summary["student_name"].count()
                                     , "School Budget": school_summary["budget"].sum()/school_summary["student_name"].count()
                                     , "Per Student Budget": (school_summary["budget"].sum()/school_summary["student_name"].count())/school_summary["student_name"].count()
                                     , "Avg Math Score": round(school_summary["math_score"].mean(),2)
                                     , "Avg Reading Score": round(school_summary["reading_score"].mean(),2)
                                     , "Percent Passing Math": percent_pass_math.map('{:.2f}%'.format)
                                     , "Perecnt Passing Reading": percent_pass_reading.map('{:.2f}%'.format)
                                     , "Overall Passing Rate":round((percent_pass_math+percent_pass_reading)/2, 2).map('{:.2f}%'.format)
                                    
                                    })
school_summary_table

```
