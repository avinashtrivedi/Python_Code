```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change the path if needed)
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# Read School and Student Data and store into Pandas Dataframes
school_data_df = pd.read_csv(school_data)
student_data_df = pd.read_csv(student_data)

# Combine the data into a single dataset
school_data_complete_df = pd.merge(student_data_df,
                                   school_data_df, how="left", 
                                   on=["school_name", "school_name"])
school_data_complete_df.head() # display few rows.
```

```python
# get different grades
school_data_complete_df["grade"].unique()
```

```python
# create a series of scores by grade levels using conditionals.

# get the data by grade
ninth_data = school_data_complete_df[school_data_complete_df["grade"]=="9th"]
tenth_data = school_data_complete_df[school_data_complete_df["grade"]=="10th"]
eleventh_data = school_data_complete_df[school_data_complete_df["grade"]=="11th"]
twelfth_data = school_data_complete_df[school_data_complete_df["grade"]=="12th"]


# group each school data by school name for average math score.

math_avg_ninth = ninth_data.groupby(["school_name"]).mean()["math_score"]
math_avg_tenth = tenth_data.groupby(["school_name"]).mean()["math_score"]
math_avg_eleventh = eleventh_data.groupby(["school_name"]).mean()["math_score"]
math_avg_twelfth = twelfth_data.groupby(["school_name"]).mean()["math_score"]

# group each school data by school name for average reading score.

reading_avg_ninth = ninth_data.groupby(["school_name"]).mean()["reading_score"]
reading_avg_tenth = tenth_data.groupby(["school_name"]).mean()["reading_score"]
reading_avg_eleventh = eleventh_data.groupby(["school_name"]).mean()["reading_score"]
reading_avg_twelfth = twelfth_data.groupby(["school_name"]).mean()["reading_score"]

```

```python
# Combine each data for average math scores by school into single dataframe
math_avg_df = pd.DataFrame({"9th": math_avg_ninth,
                           "10th": math_avg_tenth,
                           "11th": math_avg_eleventh,
                           "12th": math_avg_twelfth})
math_avg_df
```

```python
# Combine each data for average reading scores by school into single dataframe
reading_avg_df = pd.DataFrame({"9th": reading_avg_ninth,
                               "10th": reading_avg_tenth,
                               "11th": reading_avg_eleventh,
                               "12th": reading_avg_twelfth})
reading_avg_df
```

```python
# format each grade column

for grade in reading_avg_df.columns:
    reading_avg_df[grade] = reading_avg_df[grade].map('{:,.2f}'.format)
    
for grade in reading_avg_df.columns:
    math_avg_df[grade] = math_avg_df[grade].map('{:,.2f}'.format)
```
