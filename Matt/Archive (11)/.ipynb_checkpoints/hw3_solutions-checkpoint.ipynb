# 1

```python
import pandas as pd

titanic = pd.read_csv("titanic.csv")

columns = ['Pclass', 'Sex','Embarked','SibSp', 'Parch','Age','Fare','Survived']
titanic = titanic[columns]

titanic['Embarked'].fillna("S", inplace=True)
med_age = int(titanic['Age'].median())
titanic['Age'].fillna(med_age, inplace=True)

titanic['Sex'] = titanic['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
titanic['Embarked'] = titanic['Embarked'].map( {'S': 1, 'C': 2, 'Q': 3} ).astype(int)
```

```python
from sklearn.model_selection import train_test_split

X_titan = titanic.drop('Survived', axis = 1)
y_titan = titanic[['Survived']]
X_titan_train, X_titan_test, y_titan_train, y_titan_test = train_test_split(X_titan, y_titan, test_size=1/3, random_state=42)
```

## 1.1

```python
import numpy as np
from sklearn import tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

depth = np.arange(1,51,1)
dtaccuracy_train = []
dtaccuracy_test = []

for k in depth:
    # Initialize tree object.
    d_tree = tree.DecisionTreeClassifier(max_depth=k, random_state=42)
    # Fit.
    d_tree.fit(X_titan_train,y_titan_train)
    # Get test and training accuracies.
    y_titan_pred_test = d_tree.predict(X_titan_test)
    y_titan_pred_train = d_tree.predict(X_titan_train)
    # Store.
    dtaccuracy_train.append(accuracy_score(y_titan_pred_train,y_titan_train))
    dtaccuracy_test.append(accuracy_score(y_titan_pred_test,y_titan_test))

print("Best prediction accuracy: ", max(dtaccuracy_test))
print("Depth: ", depth[dtaccuracy_test.index(max(dtaccuracy_test))])

plt.plot(depth, dtaccuracy_train, label = "Training")
plt.plot(depth, dtaccuracy_test, label = "Test")
plt.xlabel("Depth")
plt.ylabel("Accuracy")
plt.title("Accuracy vs. Depth")
plt.legend()
plt.show()
```

## 1.2

### Answer:

Training accuracy increases with depth. Test accuracy increases with depth on average, then decreases. These trends are due to overfitting as depth increases.

## 1.3

```python
# Fit tree with identified depth.
d_tree = tree.DecisionTreeClassifier(max_depth=3, random_state=42)
d_tree.fit(X_titan_train,y_titan_train)
# Access and print feature importances attribute.
importances = pd.DataFrame(d_tree.feature_importances_, index=X_titan_train.columns, columns=["Importance"])
importances.sort_values(by = 'Importance', inplace=True, ascending=False)
importances.head(3)
```

# 2

## 2.1

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

rf_arru_lst_test = []
rf_arru_lst_train = []
n_estimators_lst = np.arange(1,202,10)

for k in n_estimators_lst:
    # Fit random forest.
    rf = RandomForestClassifier(max_depth=3, n_estimators=k, random_state=42)
    rf.fit(X_titan_train, y_titan_train.values.ravel())
    # Get accuracies.
    rf_y_titan_pred_test = rf.predict(X_titan_test)
    rf_y_titan_pred_train = rf.predict(X_titan_train)
    acc_random_forest_test = accuracy_score(rf_y_titan_pred_test,y_titan_test)
    acc_random_forest_train = accuracy_score(rf_y_titan_pred_train,y_titan_train)
    # Store.
    rf_arru_lst_test.append(acc_random_forest_test)
    rf_arru_lst_train.append(acc_random_forest_train)
    
print("Best prediction accuracy: ", max(rf_arru_lst_test))
print("n_estimators : ", n_estimators_lst[rf_arru_lst_test.index(max(rf_arru_lst_test))])

plt.plot(n_estimators_lst, rf_arru_lst_test, label = "Test")
plt.plot(n_estimators_lst, rf_arru_lst_train, label = "Train")
plt.xlabel('n_estimators')
plt.ylabel('Accuracy')
plt.title('n_estimators vs. Accuracy')
plt.legend()
plt.show()
```

## 2.2

### Answer:
Both training and test accuracy level off. This is because averaging predictions over more estimators decreases prediction variance.

## 2.3

```python
from sklearn.model_selection import GridSearchCV

# Create the default model.
rf = RandomForestClassifier(random_state=42)

# Grid of parameters over which to optimize.
grid = {
 'max_depth': np.arange(1,11,1),
 'n_estimators': np.arange(1,11,1)}

# Search the grid.
gridsearch = GridSearchCV(estimator=rf, param_grid=grid, cv=10,verbose=2,n_jobs=-1)
grid_result = gridsearch.fit(X_titan_train, y_titan_train.values.ravel())

# Print.
print("Best parameters: ", grid_result.best_params_)
print("Score: ", grid_result.best_score_)
```

```python
import pandas as pd

fash = pd.read_csv("fash.csv")
labels = [
    "T-shirt/top",
    "Trouser",
    "Pullover",
    "Dress",
    "Coat",
    "Sandal",
    "Shirt",
    "Sneaker",
    "Bag",
    "Ankle boot"]
fash['label'].replace(to_replace=np.arange(0,10), value=labels, inplace=True)

X_fash = fash.iloc[:, 1:]
y_fash = fash.iloc[:, 0]
```

## 3.1

```python
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Fit logistic regression model.
log_reg = LogisticRegression(penalty="none")
log_reg.fit(X_fash, y_fash)

for k in range(3):
    # Use relevant methods of fitted logistic regression object to get classifications nad predicted probabilities.
    print(log_reg.predict(X_fash.iloc[k].values.reshape(1, -1))[0])
    print(log_reg.predict_proba(X_fash.iloc[k].values.reshape(1, -1))[0])
    # Print image by reshaping data array.
    plt.imshow(X_fash.iloc[k].values.reshape(28,28), cmap = 'Greys')
    plt.show()
```

## 3.2

### Answer:
1. The addition of the scaled squared L2 norm of the parameter vector to the loss function. 
2. It promotes a simpler model by penalizing larger parameter magnitudes. A simpler model can avoid overfitting, which improves generalization.

## 3.3

```python
from numpy.linalg import norm
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings("ignore")

C = [1e-3, 1e-2, 1e-1, 1e0, 1e1, 1e2, 1e3]
cv_scores = [0]*7

for k in range(len(C)):
    # Fit logistic regression model.
    log_reg = LogisticRegression(penalty="l2", C=C[k])
    log_reg.fit(X_fash, y_fash)
    # Get mean cross-validation score.
    cv_scores[k] = np.mean(cross_val_score(log_reg, X_fash, y_fash))
    
# Print optimal C.
optimal_c = C[np.argmax(cv_scores)]
print("Optimal C:", optimal_c)

plt.plot(np.log(C), cv_scores)
plt.xlabel("log(C)")
plt.ylabel("Averaged Accuracy")
plt.title("L2 Regularization")
plt.show()
```

## 3.4

A random state was not specified for the train/test split, so answers may vary.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import plot_confusion_matrix

# Split data.
X_fash_train, X_fash_test, y_fash_train, y_fash_test = train_test_split(X_fash, y_fash, test_size=0.2, random_state=42)

# Fit model.
log_reg_opt = LogisticRegression(penalty="l2", C=optimal_c)
log_reg_opt.fit(X_fash_train,y_fash_train)

# Plot confusion matrix.
plt.figure(figsize=(12,12))
ax= plt.gca()
plot_confusion_matrix(log_reg_opt,X_fash_test,y_fash_test,ax=ax);
```

## 3.5

A random state was not specified for the train/test split, so answers may vary.

```python
# Get accuracies from confusion matrix.
from sklearn.metrics import confusion_matrix

confmatrix = confusion_matrix(y_fash_test,log_reg_opt.predict(X_fash_test))
class_accuracies = confmatrix.diagonal()/confmatrix.sum(axis=1)
print("The relatively lowest accuracy is obtained for class label \"",labels[np.argmin(class_accuracies)],".\"" , sep="")
```

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt

X_wine = datasets.load_wine(as_frame = True).data[['color_intensity', 'hue']]
y_wine = (datasets.load_wine(as_frame = True).target == 0)*1

scaler = StandardScaler()
X_wine = scaler.fit_transform(X=X_wine)

fig, ax = plt.subplots(figsize=(8,6));
scatter = ax.scatter(X_wine[:,0], X_wine[:,1],c=y_wine)
ax.legend(handles=scatter.legend_elements()[0],labels=['0','1'], fontsize=16);
ax.set_xlabel("Color Intensity (Scaled)")
ax.set_ylabel("Hue (Scaled)")
ax.set_title("Wine");
```

## 4.1

```python
from sklearn.metrics import accuracy_score
from sklearn import svm

# Fit model.
clf = svm.SVC(kernel='linear')
clf.fit(X_wine, y_wine)

# Print accuracy.
y_pred = clf.predict(X_wine)
print(accuracy_score(y_wine, y_pred))
```

## 4.2

### Answer:
The classes are not linearly separable.

## 4.3

```python
from sklearn.metrics import accuracy_score
from sklearn import svm
from sklearn.model_selection import cross_val_score

# Fit with C=1.
clf = svm.SVC(kernel='rbf', C=1)
clf.fit(X_wine, y_wine)

# Print mean accuracy.
print('C=1 Mean Accuracy:', np.mean(cross_val_score(clf, X_wine, y_wine, cv=10)))

# Fit with C=1e4.
clf = svm.SVC(kernel='rbf', C=1e4)
clf.fit(X_wine, y_wine)

# Print mean accuracy.
print('C=1e4 Mean Accuracy:', np.mean(cross_val_score(clf, X_wine, y_wine, cv=10)))
```

## 4.4

```python
import numpy as np
import matplotlib.pyplot as plt

# Create grid to evaluate model
xx = np.linspace(X_wine[:,0].min()-0.5, X_wine[:,0].max()+0.5, 30)
yy = np.linspace(X_wine[:,1].min()-0.5, X_wine[:,1].max()+0.5, 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T

# Fit the model with C=1.
clf = svm.SVC(kernel='rbf', C=1)
clf.fit(X_wine, y_wine)
# Get decision boundary and margins.
Z = clf.decision_function(xy).reshape(XX.shape)
# Get support vector indices.
sv_indices = clf.support_

# Plot.
fig, ax = plt.subplots(figsize=(8,6));
# Decision boundary and margins.
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--']);
# Scatterplot of data.
ax.scatter(X_wine[:,0], X_wine[:,1], c=y_wine)
# Highlight support vectors.
ax.scatter(X_wine[sv_indices,0], X_wine[sv_indices,1], c=y_wine[sv_indices],
           linewidths=1.5,edgecolors='r')
ax.set_title("C=1")
ax.set_xlabel("Color Intensity (Scaled)")
ax.set_ylabel("Hue (Scaled)")
ax.legend(handles=scatter.legend_elements()[0],labels=['0','1'], fontsize=16);
plt.show()

# Fit the model with C=1e4.
clf = svm.SVC(kernel='rbf', C=1e4)
clf.fit(X_wine, y_wine)
# Get decision boundary and margins.
Z = clf.decision_function(xy).reshape(XX.shape)
# Get support vector indices.
sv_indices = clf.support_

# Plot.
fig, ax = plt.subplots(figsize=(8,6));
# Decision boundary and margins.
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--']);
# Scatterplot of data.
ax.scatter(X_wine[:,0], X_wine[:,1], c=y_wine)
# Highlight support vectors.
ax.scatter(X_wine[sv_indices,0], X_wine[sv_indices,1], c=y_wine[sv_indices],
           linewidths=1.5,edgecolors='r')
ax.set_title("C=1e4")
ax.set_xlabel("Color Intensity (Scaled)")
ax.set_ylabel("Hue (Scaled)")
ax.legend(handles=scatter.legend_elements()[0],labels=['0','1'], fontsize=16);
plt.show()
```
