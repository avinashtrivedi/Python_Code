# 1 Bayesian Inference

In the cell below, we perform Bayesian inference to estimate a parameter $\mu$ using data $Y_{1}, Y_{2}, \ldots, Y_{N}$ generated from an $\mathcal{N}(\mu,1)$ distribution with unknown mean parameter $\mu$.  We assume a triangular prior distribution.

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

N=10
np.random.seed(29)
Y = stats.norm.rvs(loc=2*np.random.uniform()-1,scale=1,size=N,random_state=11)

support=np.linspace(-2,2,1000)
xx = np.linspace(-2,2,1000)

def Prior(x):
    return stats.triang.pdf(x,loc=-2,scale=4,c=0.5)

def Likelihood(x,data):
    return np.prod([stats.norm.pdf(x,val) for val in data])

def Posterior(x,data,denom):
    return Prior(x)*Likelihood(x,data)/denom

integrand = np.array([Prior(x) for x in support]) * np.array([Likelihood(x,Y) for x in support])
denominator = np.trapz(integrand,support)

pri = [Prior(x) for x in xx]
post = [Posterior(x,Y,denominator) for x in xx]
plt.plot(support,post,label="Posterior")
plt.plot(support,pri,label="Prior")
plt.legend()
plt.show()

likelihood = [Likelihood(x,Y) for x in xx]
plt.plot(support,likelihood,label="Likelihood")
plt.legend()
plt.show()
```

A student wants to compare the accuracy of sklearn's decision tree and random forest classifiers. In the cell below he performs 100 trials comparing a decision tree to a random forest with a single tree. He sees that on average, the accuracy of the decision trees was higher than that of the random forests with a single tree. He incorrectly concludes that there is a problem with sklearn's random forest classifier.

```python
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn.metrics import accuracy_score

N=100
X = np.array(X)
y = np.array(y)

np.random.seed(10)

DecisionTreeScores=np.zeros(N)
for i in range(N):
    DecisionTree = tree.DecisionTreeClassifier(max_depth=4)
    DecisionTree.fit(X,y)
    predict_DecisionTree= DecisionTree.predict(X)
    DecisionTreeScores[i] = accuracy_score(predict_DecisionTree,y)

RandomTreeScores=np.zeros(N)
for i in range(N):
    RandomTree = RandomForestClassifier(n_estimators=1,max_features=2,max_depth=4) ##n_estimators=1 to use a single tree
    RandomTree.fit(X,y)
    predict_RandomTree = RandomTree.predict(X)
    RandomTreeScores[i] = accuracy_score(predict_RandomTree,y)

print('Decision Tree Average Score: ' + str(np.mean(DecisionTreeScores)))
print('Random Tree Average Score: ' + str(np.mean(RandomTreeScores)))
```

Here we compute the distance matrix, called `distances`, associated with this data set.

```python
import numpy as np

data1 = data[np.newaxis, :, :]
data2 = data[:, np.newaxis, :]
temp = data1 - data2
distances = np.sum(temp*temp, axis=2)
```

Instead of simply using an adjacency matrix as the similarity matrix, we construct a weighted adjacency matrix $W$ with entries
$$w_{ij} = \begin{cases} 
1 - \frac{||\mathbf{x}^{(i)} - \mathbf{x}^{(j)}||}{0.1} & \text{if }||\mathbf{x}^{(i)} - \mathbf{x}^{(j)}|| \leq 0.1, \\
0 & \text{otherwise}.
\end{cases}$$

```python
W1 = (distances <= .1)
W = W1 * (1 - distances/.1)
```
