# 1. Quesion

```python
s="people can't achieve" # or "Strings are awesome!"
# Length should be 20
print("Length of s = %d" % len(s))
# First occurrence of "a" should be at index 8
print("The first occurrence of the letter a = %d" % s.index("a"))
# Number of a's should be 2

print("a occurs %d times" % s.count("a"))
# Slicing the string into bits

print("The first five characters are '%s'" % s[:5]) # Start to 5

print("The next five characters are '%s'" % s[5:10]) # 5 to 10

print("The thirteenth character is '%s'" % s[12]) # Just number 12

print("The characters with odd index are '%s'" %s[1::2]) #(0-based indexing)

print("The last five characters are '%s'" % s[-5:]) # 5th-from-last to end
# Convert everything to uppercase

print("String in uppercase: %s" % s.upper())

# Convert everything to lowercase

print("String in lowercase: %s" % s.lower())

# Check how a string starts

if s.startswith("Str"):

    print("String starts with 'Str'. Good!")

# Check how a string ends

if s.endswith("ome!"):

    print("String ends with 'ome!'. Good!")

# Split the string into three separate strings,

# each containing only a word

print("Split the words of the string: %s" % s.split(" "))
```

# 2. Question

```python
# import the library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the dataset
dataset  = pd.read_csv('Titanic.csv')
```

```python
dataset # display the content of dataset
```

```python
# survivor and non survivor dataframe
survivors = dataset[dataset['survived']=='yes']
non_survivors = dataset[dataset['survived']=='no']

# compute gender wise survivor and non_survivor status
survivors_genderwise = survivors.groupby(['gender']).size().values
non_survivors_genderwise = non_survivors.groupby(['gender']).size().values

# graph plotting
gender_categories = ['Female', 'Male']
plt.figure(figsize=(4,5))
plt.bar(range(len(survivors_genderwise)), survivors_genderwise, label='Survivors',color='g')
plt.bar(range(len(non_survivors_genderwise)), non_survivors_genderwise, bottom=survivors_genderwise, label='Non Survivors', color='r')
plt.xticks([0, 1.0], gender_categories )
plt.ylabel("Count")
plt.title("Survivors count by gender",fontsize=12)
plt.legend(loc='upper left')
plt.show()
```

# h. Is there a relationship between age and class who survived the accident?

```python
# dataframe containing age and class of passengers who survived the accident?
df_age_class_rel = data[data['survived']=='yes'][['age','class']]
df_age_class_rel
```

```python
# compute a correlation  between age and class
# its value can be in the range -1 to 1
# 1 indicates perfect positive relation
# -1 indicates perfect negative relation
# 0 means no relation.
#  Â± 0.30 to Â± 0.49 : medium relation
# source : https://www.statisticssolutions.com/pearsons-correlation-coefficient/

df_age_class_rel['age'].corr(df_age_class_rel['class'])
```

```python
# Hence age and class is having a medium relation.
```
