```python
# import the library
import numpy as np
from shapely.geometry import LineString

# load the data
# use column only for from_x;from_y;to_x;to_y which is column number 5,6,7 and 8
# skiprows=1 will skip the header information

data = np.loadtxt("travel_Times_example_2019.txt",delimiter=';',skiprows=1,usecols=[5,6,7,8])
```

```python
import matplotlib.pyplot as plt

# set the figsize
plt.figure(figsize=(10,8))

# iterate through each row
for i in range(len(data)):
    
    # get x1,y1,x2,y2
    x1,y1,x2,y2 = data[i][0],data[i][1],data[i][2],data[i][3]
    
    # plot line x1,y1,x2,y2
    plt.plot([x1,x2],[y1,y2])
    
    # get respective LineString
    a = LineString([(x1, y1), (x2, y2)])
    
    # for each line check its intersection against all other lines.
    for j in range(i+1,len(data),1):
        
        # get x and y coordinates
        xx1,yy1,xx2,yy2 = data[j][0],data[j][1],data[j][2],data[j][3]
        
        # get respective LineString
        b = LineString([(xx1, yy1), (xx2, yy2)])
        
        # get intersection
        intersection = a.intersection(b)
        
        # check if its MultiPoint or Point intersection
        # mark the intersection Point 
        if intersection.geom_type == 'MultiPoint':
            plt.plot(*LineString(intersection).xy, 'o')
        elif intersection.geom_type == 'Point':
            plt.plot(*intersection.xy, 'o')
            
# show the plot
plt.show()
```
