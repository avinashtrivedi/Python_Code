```python
# import the library
import pandas as pd
import numpy as np

# load the dataset
df = pd.read_csv('winequality.csv', sep=';')

# display few rows.
df.head()
```

```python
# display columns
df.columns
```

```python
# replace the Space in column name with '_'  i.e. underscore

df.columns = df.columns.str.replace(' ','_')
df.head()
```

```python
# Define the function.

def numeric_to_buckets(df, column_name):
    
    # Check if column name is valid.
    if column_name in df:
        
        # Compute the median
        median = df[column_name].median()
        
        # Replace column values with high and low 
        # According to the condition column value >= median
        
        df[column_name] = np.where(df[column_name]>=median,'high','low')
    else:
        print('Column name does not exist.')
```

```python
# call the function.
numeric_to_buckets(df, 'fixed_acidity')

# compute the result.
result = df.groupby('fixed_acidity').quality.mean()

# print the result.
print(result.to_string())
```
