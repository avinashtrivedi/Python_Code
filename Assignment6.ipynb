```python
import statsmodels.api as sm

# get x and y from book_data
y = book_data['average_rating']
x = book_data[['num_pages', 'ratings_count', 'text_reviews_count']]

# add constant
x = sm.add_constant(x)

# build OLS model
model = sm.OLS(y,x, missing='drop')
result = model.fit()
print(result.summary())
```

- R-squared  value of 0.029 indicate that the model explains 2.9 percent of the variation in the dependent variable
- In general, higher r-squared indicates a better fit for the model and hence more association between dependent and independent variables. 

- Also according to p-value only num_pages is statistically significant because its p-value is equal to 0.000
- But, ratings_count and text_reviews_count are not statistically significant because its p-value (0.023 and 0.916 respectively) is greater than the usual significance level of 0.05
- Regression Coefficients indicates positive/negative correlation between each independent variable and the dependent variable.
- Hence text_reviews_count and average_rating is negatively correlated (-2.497e-07), although num_pages and ratings_count is positively correlated (weak correlation) with average_rating.
