```python
# create a list to store members name
member_name = ['Charlotte','Carl','John','Kelvin','Agata','Kristian','Carina','John','Ellia','Kevin']
# Another list to store members age
member_age = [23,52,41,21,26,23,41,19,27,16]
while True:
    print('1-Search Direct Contacts')
    print('2-Search User Pool')
    print('3-Member Summary')
    print('0-Exit')
    print()
    choice = input('Choice: ')
    print()
    if choice=='1':
        name = input('Name: ') 
        print()
        print('Direct Contacts')
        print('=========================')
        
        # print the 2 members signed in before and 2 members signed in after a given member
        if name in member_name: 
            for j in range(len(member_name)):
                if name==member_name[j]:
                    if j-2>=0:
                        print(member_name[j-2])
                    if j-1>=0:
                        print(member_name[j-1])
                    if j+1<len(member_name):
                        print(member_name[j+1])
                    if j+2<len(member_name):
                        print(member_name[j+2])
        else:
            print('Cannot find member in the system')
        print()
    elif choice=='2':
        name1 = input('Name1: ') 
        name2 = input('Name2: ') 
        name3 = input('Name3: ') 
        nam_list=[name1,name2,name3]

        flag=True # to keep track if members ever signed in back to back
        # check if all three members ever signed in back to back
        for i in range(len(member_name)):
            if member_name[i] in nam_list:
                names = member_name[i:i+3]
                if set(names)==set(nam_list):
                    print('Yes these members are in direct contact')
                    flag=True
                    break
                else:
                    flag=False
        if flag==False:
            print('No Evidence in logs')
        print()
    # compute min,max and average
    elif choice=='3':
        print('Min Age: ',min(member_age))
        print('Max Age: ',max(member_age))
        print('Avg: ',sum(member_age)/len(member_age))
        print()
    elif choice=='0':
        break
    else:
        print('wrong selection,please try again.')
        print()
```
