# Import library

```python
import matplotlib.pyplot as plt
import numpy as  np
```

# Define f(x) 
- f(x) = e<sup>-x/10</sup>sin(x)

```python
def f(x):
    return np.exp(-x/10) * np.sin(x)
```

# 3. a)

```python
# creating 1000 equally spaced x value between 0 and 10pi
x = np.linspace(0, 10*np.pi, 1000)
# plot graph between x and f(x)
plt.plot(x, f(x),color='r',label='y=f(x)')
plt.xlabel('x Angle [rad]',fontsize=13)
plt.ylabel('e^(-x/10)*sin(x)',fontsize=13)
plt.legend(fontsize=13)
plt.show()
```

# 3.b)

```python
x = np.linspace(0, 10*np.pi, 1000)

# for x vs f(x) graph
plt.plot(x, f(x),color='r',label='y=f(x)')

# for x vs -f(x) graph
plt.plot(x, -f(x),color='b',label='y=-f(x)')

plt.xlabel('x Angle [rad]',fontsize=13)
plt.ylabel('e^(-x/10)*sin(x)',fontsize=13)
plt.legend(fontsize=13)
plt.show()
```
