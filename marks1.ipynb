```python
# get the inputs
n = int(input('How many marks you want to enter:'))
marks_list = []
for i in range(n):
    mark = input('Enter the marks {}: '.format(i+1))
    marks_list.append(mark)

excellent_marks = []

# print in Table format
print(55*'-')
print('| {:<5} | {:<15} | {:<25} |'.format('Mark','Letter Grade','Performance Description'))
print(55*'-')
for i in marks_list:
    
    # get rounded marks 
    mark = int(round(float(i) + .1))
    
    if 90<=mark<=100:
        
        # store Excellent marks
        excellent_marks.append(float(i))
        print('| {:<5} | {:<15} | {:<25} |'.format(i,'A+','Excellent'))
        print(55*'-')
    elif 85<=mark<=89:
        # store Excellent marks
        excellent_marks.append(float(i))
        print('| {:<5} | {:<15} | {:<25} |'.format(i,'A','Excellent'))
        print(55*'-')
    elif 80<=mark<=84:
        # store Excellent marks
        excellent_marks.append(float(i))
        print('| {:<5} | {:<15} | {:<25} |'.format(i,'A-','Excellent'))
        print(55*'-')
    elif 77<=mark<=79:
        print('| {:<5} | {:<15} | {:<25} |'.format(i,'B+','Good'))
        print(55*'-')
    elif 73<=mark<=76:
        print('| {:<5} | {:<15} | {:<25} |'.format(i,'B','Good'))
        print(55*'-')
    elif 70<=mark<=72:
        print('| {:<5} | {:<15} | {:<25} |'.format(i,'B-','Good'))
        print(55*'-')
    elif 67<=mark<=69:
        print('| {:<5} | {:<15} | {:<25} |'.format(i,'C+','Satisfactory'))
        print(55*'-')
    elif 63<=mark<=66:
        print('| {:<5} | {:<15} | {:<25} |'.format(i,'C','Satisfactory'))
        print(55*'-')
    elif 60<=mark<=62:
        print('| {:<5} | {:<15} | {:<25} |'.format(i,'C-','Satisfactory'))
        print(55*'-')
    elif 57<=mark<=59:
        print('| {:<5} | {:<15} | {:<25} |'.format(i,'D+','Marginal'))
        print(55*'-')
    elif 53<=mark<=56:
        print('| {:<5} | {:<15} | {:<25} |'.format(i,'D','Marginal'))
        print(55*'-')
    elif 50<=mark<=52:
        print('| {:<5} | {:<15} | {:<25} |'.format(i,'D-','Marginal'))
        print(55*'-')
    elif 0<=mark<=49:
        print('| {:<5} | {:<15} | {:<25} |'.format(i,'F','Unsatisfactory'))
        print(55*'-')
    else:
        print('| {:<5} | {:<15} | {:<25} |'.format(i,'Not Possible','Not Possible'))
        print(55*'-')     
print('The Excellent marks are: ',excellent_marks)

sum_sq = sum([i*i for i in excellent_marks])
print('The sum of squares is:',sum_sq)
```
