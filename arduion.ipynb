```python
#importing library
import numpy as np
import matplotlib.pyplot as plt

theta1 = np.linspace(0, 10, 100) #generate 100 evenly spaced numbers between 0-10
r1 = np.sqrt(2.0) # sample radius for circle-1
r2 = np.sqrt(4.0) # sample radius for circle-2

#generating x1,y1 for circle-1 values to plot the circle,as data in not given in question.
xCoords1 = r1*np.cos(theta1)
yCoords1 = r1*np.sin(theta1)
theta2 = np.linspace(0, 10, 100) ##generate 100 evenly spaced numbers between 0-10

#generating x2,y2 for circle-2 values to plot the circle,as data in not given in question.
xCoords2 = r2*np.cos(theta2)
yCoords2 = r2*np.sin(theta2)
#plot function as mentioned in question.
def plotCurves(xCoords1,yCoords1,legend1, xCoords2,yCoords2,legend2):
    try:
        plt.plot(xCoords1, yCoords1,label=legend1)
        plt.plot(xCoords2, yCoords2,label=legend2)
        plt.xlim(-5,5)
        plt.ylim(-5,5)
        plt.title('drawing two circles', fontsize=15)
        plt.xlabel("x axis",fontsize=15)
        plt.ylabel("y axis",fontsize=15)
        plt.legend(fontsize=15)
        plt.savefig("plot_circle.png", bbox_inches='tight')
        plt.show()
    except:
        print('input sequence is wrong.')
```

```python
plotCurves(xCoords1,yCoords1,'circle-1', xCoords2,yCoords2,'circle-2')
```

```python
import numpy as np
import matplotlib.pyplot as plt
  
x = np.linspace( -0.7 , 0.7 , 150 )
y = np.linspace( -0.7 , 0.7 , 150 )
  
a, b = np.meshgrid( x , y )
  
C = a ** 2 + b ** 2 - 0.2
  
figure, axes = plt.subplots()
  
axes.contour( a , b , C , [0] )
axes.set_aspect( 1 )
  
plt.title( 'Center-Radius form Circle' )
plt.show()
```
