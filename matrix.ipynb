```python
# calculate the number of years between each subsequent visit
agediff = age.diff(axis = 'columns') 

# drop the column containing NaN values
agediff.dropna(axis = 'columns',inplace=True)
agediff
```

```python
# create a "flattened" version of the above 500x5 matrix
m = np.matrix(agediff) 
flatten_matrix = m.flatten()
flatten_matrix
```

```python
# convert matrix to numpy array
flatten_matrix = flatten_matrix.A1
```

```python
# plot the histogram
plt.hist(flatten_matrix,bins=10,edgecolor="k")

# compute average and Standard deviation
avg = flatten_matrix.mean()
sd = flatten_matrix.std()

# annotations to mark the average and standard deviation 
plt.annotate('Average = (%.1f)' % (avg),
             (2,300))
plt.annotate('Standard deviation = (%.1f)' % (sd),
             (2,200))

# show the plot
plt.show()
```
