```python
print('welcome to the Alien Bug Containment Game')
print("Trap the Alien Bug before your ship reaches Earth.")
print("Before searching for the alien bug you will need: helmet,crate, tranquilizer gun, book and shield.")
inventory = ['Mission Control','Armory','Firing Range','Storage bay','Life Support','Library',
                 'Exercise Room','Storage Unit']

# create dictionary to store mapping data
item={'Armory':'tranquilizer gun',
      'Firing range':'shield',
      'Storage bay':'Alien bug!!',
      'Library':'book',
      'Exercise room':'helmet',
      'Storage unit':'crate',
      'Life support':'Tranquilizer gun'
}

mapping = {
    'Mission Control North':'Armory',
    'Mission Control South':'Exercise room',
    'Mission Control East':'Life support',
    'Mission Control West' : 'Library',
    'Armory East':'Firing range',
    'Armory South':'Mission Control',
    'Firing range West' :'Armory',
    'Storage bay South':'Life support'   ,
    'Life support West':'Mission Control',
    'Life support North':'Storage bay',
    'Storage unit West':'Exercise room',
    'Exercise room North':'Mission Control',
    'Exercise room East' : 'Storage unit',
    'Library East':'Mission Control'
}

dir_int ={
    0:'North',
    1:'South',
    2:'East',
    3:'West'
}

# to store picked items
items=[]
```

```python
# function to go into next room using current room and direction.
def get_cur_room(inventory,direction):
    if (inventory+' '+direction) in mapping:   
        cur_room = mapping[inventory+' '+direction]
        print('Welcome to ',cur_room)
        if cur_room in item:
            print('Available item is ',item[cur_room])
            try:
                action = input('Pick the item or leave, yes/no: ')
            except:
                print('invalid selection input')
            if action =='yes':
                if item[cur_room] not in items:
                    items.append(item[cur_room])
                    print('Item has been Saved.')
                else:
                    print('Item ',item[cur_room],' is already Saved')
                print(50*'*') # print * 50 times
            else: 
                print('Item has been Discarded.')
                print(50*'*')
        return cur_room
    else:
        print('Invalid Direction')
        return inventory
```

```python
CurrentRoom = 'Mission Control'
flag=True # flag to continue or exit the game.
print('You are in the Mission Control now')
while flag:
    print('Please choose the direction:')
    print('North: 0')
    print('South: 1')
    print('East:  2')
    print('West:  3')
    try:
        d = int(input())
        CurrentRoom = get_cur_room(CurrentRoom,dir_int[d])
    except:
        print('invalid input')
    flag = input('Do you want to continue? yes/no: ')
    try:
        flag={'yes':True,'no':False}[flag]
    except:
        print('Wrong input')
print('List of items: ',items)
```
