```python
def func2(n):
    
    # list to store students detail
    student_details = []
    
    # iterate through the loop n times
    # to store the details for n student.
    for i in range(n):
        print('Enter details for student Number:',i+1)
        try:
            sname = input('Please enter student Name: ')
            sgrade = int(input('Please enter student Grade: '))
            print()
            if sname == '': # if name is empty
                print('Please enter a valid name')
                
            # store name and grade as a tuple of the form (name,grade)
            student_details.append((sname,sgrade))
        except:
            print('Wrong inputs.')
            
    print('Name of student with Maximum Exam Grade\n')
    print("{:<20} {:<5}".format('Name','Grade'))
    print(30*"-") # print '-' , 30 times for better print formatting
    
    # get the maximum grade
    max_grade = max(student_details,key=lambda x:x[1])
    
    # print name of all the students having grade equal to maximum grade
    for name,grade in student_details:
        if grade == max_grade[1]:
            print("{:<20} {:<5}".format(name,grade))
```

```python
func2(5)
```
