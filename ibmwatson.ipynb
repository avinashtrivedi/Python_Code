```python
# import library
from ibm_watson import NaturalLanguageUnderstandingV1
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator
from ibm_watson.natural_language_understanding_v1 import Features, EntitiesOptions

# Authenticating the API access
authenticator = IAMAuthenticator('bKaaWQVwtE807KO3VPK-1QzsCLsa6oABTLtTuT0BjDg9')  # api key
natural_language_understanding = NaturalLanguageUnderstandingV1(
    version='2021-03-25',
    authenticator=authenticator
)

# create function
def extractEntities(url, source):
    
    # base URL for natural_language_understanding
    natural_language_understanding.set_service_url('https://api.eu-gb.natural-language-understanding.watson.cloud.ibm.com/instances/e78bd03a-7e97-4fba-ba3d-22cfc10e23a6')
    response = natural_language_understanding.analyze(
        url=url,
        features=Features(entities=EntitiesOptions(sentiment=True))).get_result()
    # response is a dictionary,Please print response to check its structure
    
    # list to store dictionary for each entity
    list_dict = []
    
    # go through the response dictionary and extract the desired field.
    for i in range(len(response['entities'])):
        temp_dict={}
        temp_dict['entity'] = response['entities'][i]['text']
        temp_dict['url'] = response['retrieved_url']
        temp_dict['source'] = source
        temp_dict['relevance'] = response['entities'][i]['relevance']
        temp_dict['sentiment'] = response['entities'][i]['sentiment']['score']
        list_dict.append(temp_dict)
    return list_dict
```
