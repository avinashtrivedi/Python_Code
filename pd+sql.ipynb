# Q.a

# i

```python
# import the library
import pandas as pd
import numpy as np

# read the dataset
ship = pd.read_csv('ship.csv',na_values='.')
```

```python
ship # display the dataframe
```

# ii

```python
# rename the column
ship = ship.rename(columns={"T": "types", "A": "c_years", "P": "o_periods", "MS": "s_months", "Y": "incidents"})
```

```python
ship # display
```

# iii

```python
shipgroup = ship.groupby(['types', 'o_periods'])[['s_months', 'incidents']].mean().round()
```

```python
shipgroup
```

# iv

```python
# for type as 1 and o_periods as 1 
# s_months average is
shipgroup.loc[1,1][0]
```

```python
# for type as 1 and o_periods as 1 
# incidents average is
shipgroup.loc[1,1][1]
```

```python
# Apply the above concept to fill th NaN values.
# x['types'] and x['o_periods'] will give the corresponding types and o_periods value.

ship['s_months'] = ship.apply(lambda x: shipgroup.loc[x['types'],x['o_periods']][0] if pd.isnull(x['s_months']) else x['s_months'], axis=1)

ship['incidents'] = ship.apply(lambda x: shipgroup.loc[x['types'],x['o_periods']][1] if pd.isnull(x['incidents']) else x['incidents'], axis=1)
```

```python
ship
```

# v

```python
# save the incidents variable data
Y = pd.DataFrame(ship['incidents'])
```

```python
X.head()
```

```python
# Convert categorical variable into dummy variables
X = pd.get_dummies(X,columns = ['types', 'c_years','o_periods'])
```

# iii

```python
# Apply log-transformation
# values rounded to 2 decimal places.

log_s_months = round(np.log(ship['s_months']),2)
```

```python
# Add 'log_s_months' column in both the dataframe ship and X

ship['log_s_months'] = log_s_months
X['log_s_months'] = log_s_months
```
