```python
class Utility_bill(object):

    def __init__(self, cus_name, cus_address):
        self._cus_name = cus_name
        self._cus_address = cus_address
        self._total = 0

    def calculate_charge(self):
        raise NotImplementedError("Not implemented")

    def display_bill(self):
        raise NotImplementedError("Not implemented")

class Water_bill(Utility_bill):

    def __init__(self,cus_name, cus_address):
        super().__init__(cus_name, cus_address)    
        self._No_of_Gallons = 0

    def calculate_charge(self):
        self._No_of_Gallons = float(input("Number of gallons of water used? "))
        print()
        # if _No_of_Gallons is <= 6000 then charge is $0.005 per gallon
        if self._No_of_Gallons <= 6000:
            self._total = 0.005 * self._No_of_Gallons
        
        # if _No_of_Gallons > 6000 then for first 6000 gallons charge is $0.005
        # and for remaining gallons charge is $0.007 per gallon
        else:
            self._total = 0.005 * 6000 + 0.007 * (self._No_of_Gallons-6000)
  
    def display_bill(self):
        print('*********************** Water Bill ***********************')
        print("{:<30} {:<40} {:<20} {:<20}".format('Name','Address','Gallons used','total charge'))
        print("{:<30} {:<40} {:<20} {:<20}".format(self._cus_name,self._cus_address,self._No_of_Gallons,self._total))
        print()
    
class Electricity_bill(Utility_bill):
  
    def __init__(self, cus_name, cus_address):
        super().__init__(cus_name, cus_address)     
        self._Kilo_Watt_Hours = 0

    def calculate_charge(self):
        self._Kilo_Watt_Hours = float(input("Enter kilowatt hours used: "))
        print()
        
        # if _Kilo_Watt_Hours is <= 500 then charge is $0.12 per kWh
        if self._Kilo_Watt_Hours <= 500:
            self._total = 0.12 * self._Kilo_Watt_Hours
        
        # if _Kilo_Watt_Hours > 500 then for first 500 kWh charge is $0.12 per kWh
        # and for remaining _Kilo_Watt_Hours charge is $0.15 per kwh
        else:
            self._total = 0.12 * 500 + 0.15 * (self._Kilo_Watt_Hours-500)
  
    def display_bill(self):
        print("********************Electricity Bill*********************")
        print("{:<30} {:<40} {:<20} {:<20}".format('Name','Address','Gallons used','total charge'))
        print("{:<30} {:<40} {:<20} {:<20}".format(self._cus_name,self._cus_address,self._Kilo_Watt_Hours,self._total))
        print()
def main():
    c_name = input("Enter the customer name: ")
    c_address = input("Enter the customer address: ")
    while True:
        try:
            print("Enter 1 for water bill")
            print("Enter 2 for electricity bill")
            print("Enter 3 to exit.")
            choice = int(input())
            if choice == 1:
                water_obj = Water_bill(c_name, c_address)
                water_obj.calculate_charge()
                water_obj.display_bill()
            elif choice == 2:
                electricity_obj = Electricity_bill(c_name, c_address)
                electricity_obj.calculate_charge()
                electricity_obj.display_bill()
            elif choice==3:
                break
            else:
                print('wrong input')
        except:
            print('Wrong input,Please try again.')

if __name__ == "__main__":
    main()
```
