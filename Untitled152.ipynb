```python
Employee = []

def cls():
    print("\n" * 50)

def employeeFormatedInfo(name, ssn, phone, email, salary):
    print("")
    print("-------------------------- {0:s} -----------------------".format(name))
    print("SSN: {0:s}".format(ssn))
    print("Phone: {0:s}".format(phone))
    print("Email: {0:s}".format(email))
    print("Salary: {0:s}".format(salary))
    print("--------------------------------------------------------")
    print("")

def viewEmployeeInfo():
    cls()
    print("--------------------------------------------------------")
    print("")
    print("            view all employees in the system            ")
    print("")
    print("--------------------------------------------------------")
    print("")
    if (len(Employee) == 0):
        print("----------------------------------------------------------")
        print("                 no employees in the list                 ")
        print("----------------------------------------------------------")
    else:
        for i in range(0, len(Employee)):
            line = Employee[i].split(",")
            employeeFormatedInfo(line[0], line[1], line[2], line[3], line[4])
    try:
        option = int(input("Enter 0 to exit or any key to continue. "))
        if option == 0:
            cls()
    except:
        cls()

def addEmployeeToList():
    cls()
    print("---------------------------------------------------------")
    print("")
    print("                add employee information                 ")
    print("")
    print("---------------------------------------------------------")
    print("")
    try:
        name = input("Employee's name: ")
        ssn = input("Employee's SSN: ")
        phone = input("Employee's phone number: ")
        email = input("Employee's email:  ")
        salary = input("Employee's salary: ")
        line = name + ',' + ssn + ',' + phone + ',' + email + ',' + salary
        Employee.append(line)

    except:
        cls()
        addEmployeeToList()
    print("")
    print("employee information has been successfully added to the list.")
    print("")
    print("----------------------------------------------------------------")
    print("")
    try:
        option = int(input("enter 0 to return to main menu, or any other number to add a new employee: "))
        if option == 0:
            cls()
        else:
            cls()
            addEmployeeToList()
    except:
        cls()
        addEmployeeToList()


def searchEmployee():
    ssn = input("Enter employees SSN to search: ")

    for employee in Employee:
        employee = employee.split(",")
        if ssn == employee[1]:
            employeeFormatedInfo(employee[0], employee[1], employee [2], employee[3], employee[4])
            break
    else:
      print("There is no Employee associated with SSN: {}".format(ssn))

    option = int(input("Enter 0 to return to main menu, or any other number to search another employee: "))

    if option == 0:
        cls()
    else:
        cls()
        searchEmployee()


def updateEmployeeInfo():
    ssn = input("Enter SSN for employee you wish to update: ")
    for i in range(len(Employee)):
        if ssn == Employee[i].split(',')[1]:
            name = input("Update employee name: ")
            ssn = input("Update employee SSN: ")
            phone = input("Update employee phone number: ")
            email = input("Update employee email: ")
            salary = input("Update employee salary: ")
            line = name + ","+ssn+","+phone+","+email+","+salary
            Employee[i] = line
            print("Employee information has been updated successfully.")
            break
    else:
        print("There are no employees associated with SSN: {}".format(ssn))
    option = int(input("Enter 0 to return to main menu or any other number to update another employees information: "))
    if option == 0:
        cls()
    else:
        cls()
        updateEmployeeInfo()

#THIS IS WHERE I WOULD LIKE TO INPUT THE FUNCTION TO EXPORT EMPLOYEE DATA

#THIS IS WHERE I WOULD LIKE TO INPUT THE FUNCTION TO IMPORT EMPLOYEE DATAï»¿

def printoptions():
    print("--------------Employee Managment System-------------------\n")
    print("There are ( {0:2d} ) employees in the system.".format(len(Employee)))
    print("\n----------------------------------------------------------")
    print("1. View all employees \n")
    print("2. Add a new employee \n")
    print("3. Search employee by SSN \n")
    print("4. Update employee information \n")
    try:
        answer = int(input("Please enter your selection: "))
    except ValueError:
        print("not a number")
        return 100
    print("")
    print("-------------------------------------------------------------")
    print("")
    return answer


while True:
    cls()
    mode = printoptions()
    if mode == 1:
        cls()
        viewEmployeeInfo()
    if mode == 2:
        cls()
        addEmployeeToList()
    if mode == 3:
        cls()
        searchEmployee()
    if mode == 4:
        cls()
        updateEmployeeInfo()
```
