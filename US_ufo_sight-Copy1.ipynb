```python
# import the library
# Read the csv file, Where each row is a separate dictionary
# and key is column of the csv file.

import csv
filepath = "ufo-sightings.csv"
ufosightings = [] 
with open(filepath, 'r') as csvfile: 
    reader = csv.DictReader(csvfile) #load and read into a DictReader
    for i in reader:
        ufosightings.append(i)
```

```python
# get the data for 'US' i.e. US sightings
sightings_us = [row for row in ufosightings if row['country'] == 'us']
```

```python
# create a list to store fresh dictionaries with key as datetime and state
fball = [] 

# iterate through the sightings_us
for row in sightings_us:
    
    # get the data for shape='fireball' having float of 'duration (seconds)' > 10
    if row["shape"] == "fireball" and float(row['duration (seconds)'])>10:
        
        # store datetime,state for each row as dictionary items.
        fball.append({'datetime':row['datetime'],'state':row['state']})
```

```python
# output printing
print('{:<25} {:<2} {:<5}'.format('datetime','|','state'))
print('***********************************')
for row in fball:
    print('{:<25} {:<2} {:<5}'.format(row['datetime'],'|',row['state']))
    print('----------------------------------')
```
