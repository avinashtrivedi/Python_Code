```python
def is_balance(X):
    total = 0
    
    # compute the sum of components of a cluster
    for i in X:
        total = total + i
    
    # check if sum is odd/even 
    # if its odd, it means, it can't be balanced
    if total%2 != 0:
        return False
    
    # Now the target is to sequentially add the cluster components
    # and check if its equal to (total/2), if it is then Pan can be balanced
    # if at any point of sequential addition, the HalfTotal becomes greater than 
    # (total/2) ,that means Pan can't be balanced.
    
    HalfTotal=0
    for i in X:
        HalfTotal = HalfTotal + i # sequential addition
        if HalfTotal == (total/2):
            return True
        elif HalfTotal > (total/2):
            return False
```

```python
X = [1, 1, 1, 2, 1]
is_balance(X) # call the function to check
```
