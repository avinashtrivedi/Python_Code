```python
# this function receive json data as a argument
def BikeIdName(jdata):
    bike_details = jdata['data']['bikes']
    bike_id_name =[] # Store bike_id and name
    for  i in bike_details:
        # append bike_id and name in list bike_id_name
        bike_id_name.append((i['bike_id'],i['name']))
    return bike_id_name # return bike_id_name list
```
