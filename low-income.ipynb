# Load the dataset

```python
import pandas as pd
dataCleansing_DF = pd.read_csv('K303_DataPreparation_ClassExample.csv')
dataCleansing_DF
```

# customer column

```python
# remove rows with blanks
dataCleansing_DF.dropna(subset=['Customer'], inplace=True)
dataCleansing_DF
```

```python
# remove rows with 9999
dataCleansing_DF.drop(dataCleansing_DF[dataCleansing_DF.Customer==9999].index,inplace=True)
dataCleansing_DF
```

```python
# remove duplicates
dataCleansing_DF.drop_duplicates(subset=['Customer'],inplace=True)
dataCleansing_DF
```

```python
# Make 'customer' as row index
dataCleansing_DF.set_index('Customer', inplace=True)
dataCleansing_DF
```

# SSN

```python
# remove extra spaces and other non-digit characters
new_ssn = [] # store ssn without space and non-digit characters
for ssn in list(dataCleansing_DF.SSN):
    temp_ssn = ''
    for char in ssn:
        if char.isdigit() == True: # check for digit
            temp_ssn = temp_ssn + char
    new_ssn.append(temp_ssn)
```

```python
dataCleansing_DF['SSN'] = new_ssn # assign fresh ssn without space and non-digit characters
dataCleansing_DF
```

```python
# remove duplicate
dataCleansing_DF.drop_duplicates(subset=['SSN'],inplace=True)
dataCleansing_DF
```

```python
# remove rows that donot have 9 digits SSN
dataCleansing_DF.drop(dataCleansing_DF[dataCleansing_DF.SSN.str.len()!=9].index, inplace=True)
dataCleansing_DF
```

```python
# just for concept
dataCleansing_DF.SSN.str.len()
```

```python
# just for concept
a=list(dataCleansing_DF.SSN)[0]
a
```

```python
# just for concept,this is how we are formatting SSN
a[:3] +'-'+a[3:5]+'-'+a[5:]
```

```python
dataCleansing_DF.SSN = dataCleansing_DF.SSN.apply(lambda x : x[:3] +'-'+x[3:5]+'-'+x[5:])
dataCleansing_DF
```

# BirthDate

```python
dataCleansing_DF.dtypes # data type of birthdate is object
```

```python
# first replace '|' with '-'
dataCleansing_DF.Birthdate = dataCleansing_DF.Birthdate.apply(lambda x : x.replace('|','-'))
# now change the datatype to datetime
dataCleansing_DF['Birthdate'] = pd.to_datetime(dataCleansing_DF['Birthdate'])
# check the data type again for confirmation,whether its changed or not
dataCleansing_DF.dtypes
```

```python
# remove rows with 9999
dataCleansing_DF.drop(dataCleansing_DF[dataCleansing_DF.Income==9999].index,inplace=True)
dataCleansing_DF
```
