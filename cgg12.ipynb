```python
# import the library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the dataset
df = pd.read_csv('vgsales.csv')
```

```python
# filter the dataframe for Publisher as Activision and time period 2000 through 2016 
df = df[(df['Publisher']=='Activision') & (df['Year']>=2000) & (df['Year']<=2016)].sort_values(by='Year')

# filter the dataframe for Genre as Sports
df = df[(df['Genre'] == 'Sports')]

# groupby 'year' to get the count of number of publication
t1 = df.groupby('Year')['Year'].count()

# since t1 is of type Series, convert it to dataframe
df1 = pd.DataFrame(t1)

# change the column name
df1.columns = ['new_value']

# shift to get old_values
df1['old_value'] = df1['new_value'].shift()

# delete NaN rows
df1.dropna(inplace=True)

# get percent_change
df1['percent_change'] = (df1['new_value'] - df1['old_value'])*100/df1['old_value']

# plot the graph
plt.plot(df1.index,df1.percent_change,label='sports')

# set the title
plt.ylabel('% Change Year over Year')

# show the legend
plt.legend()

# show the plot
plt.show()
```
