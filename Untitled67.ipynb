```python
import csv
```

```python
def read_file(fp):
    '''return a list of list i.e. list of rows'''

    myreader = csv.reader(fp)
    
    # skip unnecessary rows i.e. header, empty rows etc.
    for i in range(4):
        next(myreader)
        
    datalist = list(myreader)
    datalist = [i for i in datalist if i[0]!='']
    return datalist
```

```python
fp = open("immigration.csv")
L = read_file(fp)

# next(reader,None) # skips a line, such as a header line
# for line in reader: 
```

```python
SumUS=0
SumState=0
for col in L:
    Temp_Population = col[1].replace(',', '')
    Temp_Population = Temp_Population.replace('<','')
    Population = float(Temp_Population)
    
    if col[0]=='U.S.':
        SumUS = Population
    else:
        SumState = SumState + Population
return SumUS,SumState
```

```python
US_Unauthorized_immigrant = L[0][2]
US_Unauthorized_immigrant = float(US_Unauthorized_immigrant.replace('%', ''))
List_Of_Greater_Immigration = []
for col in L[1:]:
    StateImmigration = float(col[2].replace('%', ''))
    if StateImmigration > US_Unauthorized_immigrant:
        List_Of_Greater_Immigration.append(col[0])
return List_Of_Greater_Immigration
```

```python
def get_largest_states(L):
    # define a list
    greatestStateLists = []
    # get the nauthorized immigrant % of population of U.S.
    unauthor_Immegration = float(L[0][2].replace('%',''))
    
    # loop through rest of the rows except the "U.S." row
    for row in L[1:]:
        # get the float valur of the present at index 2
        value = float(row[2].replace('%',''))
        # check for the greatest unauthorized immigrant % of population with respect
        # to U.S. unauthorized immigrant % of population       
        if value > unauthor_Immegration:
            # Append state to the list which ever having % greater than
            # U.S. % of population
            greatestStateLists.append(row[0])
    # sort the list greatestStateLists()
    greatestStateLists = sorted(greatestStateLists)
    # return the list
    return greatestStateLists
```
