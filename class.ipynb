```python
class Student(object):
    
    # constructor of a student class
    def __init__(self):
        
        # get email id and name
        self.email = self.getEmail()
        self.name = self.getName()
              
    # function to get the student and instructor name
    def getName(self):
        
        # this while loop will execute until program will receive a valid name
        while True:
            self.name = input('Enter Name: ')
            if self.name.replace(' ','').isalpha():  # replace space and check for alphabet letters
                break
            else:
                print('Wrong input,Please try again.')
        return self.name
    
    # function to get the student id
    def getId(self):
        while True:
            self.id = input('Enter ID: ')
            # check for digits and length
            if self.id.isnumeric() and len(self.id)<=7:
                break
            else:
                print('Wrong input,Please try again.')
        return self.id
    
    # function to get student's program of study
    def getProgram(self):
        while True:
            self.program = input('Enter program of study: ')
            # remove spaces if any and check for alphabet letters
            if self.program.replace(' ','').isalpha():
                break
            else:
                print('Wrong input,Please try again.')
        return self.program
    
    # function to get email id
    def getEmail(self):
        while True:
            self.email = input('Enter Email id: ')
            if self.email.count('@')==1 and self.email.count('.')>=1:
                temp_email = self.email.replace('@','') # remove @
                temp_email = temp_email.replace('.','') # remove .  (e.g.  hello@gmail.com)
                if temp_email.isalnum(): # check for alphaneumeric characters
                    break
                else:
                    print('Wrong input,Please try again.')
            else:
                print('Wrong input,Please try again.')
        return self.email


# inheritance reference: https://www.geeksforgeeks.org/inheritance-in-python/
class Instructor(Student): # Here Student class is inherited
    
    #  function to get instructor id
    def getIId(self):
        while True:
            self.iid = input('Enter Instructor ID: ')
            # check for digits and length
            if self.iid.isnumeric() and len(self.iid)<=5:
                break
            else:
                print('Wrong input,Please try again.')
        return self.iid
    
    # function to get last institution attended
    def getLastInstitution(self):
        while True:
            self.institution = input('Enter name of last institution attended: ')
            if self.institution.replace(' ','').isalpha():
                break
            else:
                print('Wrong input,Please try again.')
        return self.institution
    
    # function to get last degree
    def getHighestDegree(self):
        while True:
            self.degree = input('Last degree received: ')
            if self.degree.replace(' ','').isalpha():
                break
            else:
                print('Wrong input,Please try again.')
        return self.degree


# validator class
class Validator():
    
    # function to validate objects
    def validate(self,obj):
        c= str(type(obj)).split('.')[1].split("'")[0]
        if c=='Student':
            
            # remove @ and . from email id before validating
            myemail = obj.email.replace('@','') 
            myemail = myemail.replace('.','')
            
            if obj.name.replace(' ','').isalpha() and myemail.isalnum() and obj.id.isnumeric() \
            and obj.program.replace(' ','').isalpha():
                return 'passed'
            else:
                return 'failed'

        elif c=='Instructor':
            # remove @ and . from email id before validating
            myemail = obj.email.replace('@','')
            myemail = myemail.replace('.','')
            if obj.name.replace(' ','').isalpha() and myemail.replace('@','').isalnum() \
                and obj.iid.isnumeric() and obj.institution.replace(' ','').isalpha()\
                and obj.degree.replace(' ','').isalpha():
                return 'passed'
            else:
                return 'failed'
```

```python
def displayInformation(data):
    # {:<10} : this is used for proper formatted printing
    print("{:<10} {:<20} {:<30} {:<40}".format('id','Name','Email','program of study'))
    for obj in data:
        c= str(type(obj)).split('.')[1].split("'")[0]     
        if c=='Student':
            print("{:<10} {:<20} {:<30} {:<40}".format(obj.id,obj.name,obj.email,obj.program)) 
    print()
    print('-----------------------------------------------------------------------------------')
    print()
    print("{:<10} {:<20} {:<30} {:<40} {:<30}".format('id','Name','Email','last institution','Degree'))
    for obj in data:
        c= str(type(obj)).split('.')[1].split("'")[0]
        if c=='Instructor':
            print("{:<10} {:<20} {:<30} {:<40} {:<30}".format(obj.iid,obj.name,obj.email,obj.institution,obj.degree))

def main():         
    data=[]
    # create object of Validator class
    valid_obj = Validator()
    while True:
        print('-----------------------------')
        print('Collect Student details 1:')
        print('Collect Instructor details 2:')
        print('Stop the program 3:')
        x = int(input('Enter your choice: '))
        print()
        if x==3:
            break
        elif x==1:
            print('---------- Student Details -------------------')
            obj=Student()
            obj.getId()
            obj.getProgram()
        elif x==2:
            print('-----------Instructor Details ----------------')
            obj=Instructor()
            obj.getIId()
            obj.getLastInstitution()
            obj.getHighestDegree()

        if valid_obj.validate(obj)=='passed':
            print('****Validation passed****')
            print()
            data.append(obj)
        else:
            print('****Validation failed*****')
            print()
    return data

if __name__ == '__main__':
    object_list = main() # call the main function to print menu
    displayInformation(object_list) # display the details.   
```
