```python
# import the library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# load the data
gapminder = pd.read_csv('gapminder.csv')
```

```python
# use units and estimator to plot lines for each observations.

fgrid = sns.relplot(
        data=gapminder,
        x="year",
        y="gdpPercap",
        hue="continent",
        col="continent",
        kind="line",
        col_wrap=3,
        units="country",
        estimator=None)

# set yscale
fgrid.set(yscale="log")

# show the graph
plt.show(fgrid)
```
