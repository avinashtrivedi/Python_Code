```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
# derivative function
def f(t, y):
    dydt = np.cos(y*t)
    return dydt

# define initial conditions and time spans
yinit = [0,.5,1,1.5]
tspan = np.linspace(0, 5, 1000)
```

```python
sol = solve_ivp(f,[tspan[0], tspan[-1]], yinit, t_eval=tspan, rtol = 1e-5)
# Plot the graph
# create a subplot of 2x2
# use one subplot for 1 graph
fig, ax = plt.subplots(2,2,figsize=(7,7))
# for y=0
ax[0,0].plot(sol.t,sol.y[0], 'k-',label='y='+ str(yinit[0]),color='g')
ax[0,0].legend(loc='best')
# for y=.5
ax[0,1].plot(sol.t,sol.y[1], 'k-',label='y='+ str(yinit[1]),color='r')
ax[0,1].legend(loc='best')
# for y=1
ax[1,0].plot(sol.t,sol.y[2], 'k-',label='y='+ str(yinit[2]),color='y')
ax[1,0].legend(loc='best')
# for y=1.5
ax[1,1].plot(sol.t,sol.y[3], 'k-',label='y='+ str(yinit[3]),color='b')
ax[1,1].legend(loc='best')
plt.show()
```
