```python
# get room type
def get_roomtype():
    while True:
        try:
            room_type = int(input('Room type: '))
            if 1<=room_type<=3:
                break
            else:
                print('Wrong Room type, Please try again.')
        except:
            print('Wrong Room type, Please try again.')
    return room_type
```

```python
# get number of guest
def get_no_of_guest():
    while True:
        try:
            no_of_guest = int(input('Number of guests: ')) 
            if no_of_guest>=1:
                break
            else:
                print('Wrong Number of guests:, Please try again.')
        except:
            print('Wrong Number of guests:, Please try again.')
    return no_of_guest
```

```python
# get number of days
def get_no_of_day():
    while True:
        try:
            no_of_days = int(input('Number of Days: ')) 
            if no_of_days>=1:
                break
            else:
                print('Wrong Number of Days')
        except:
            print('Wrong Number of Days')
    return no_of_days
```

```python
# get number of concession tickets
def get_concession():
    while True:
        try:
            no_of_concession = int(input('Number of concession tickets included (under 18, over 65 or student): '))
            if no_of_concession>=0:
                break
            else:
                print('Wrong input, pleae try again')
        except:
            print('Wrong input, pleae try again')
    return no_of_concession
```

```python
# get additional benefits
def get_additionals():
    while True:
        try:
            special_view = input('Special view: ')
            if special_view in ('Y','N'):
                break
            else:
                print('Wrong input, Please try again.')
        except:
             print('Wrong input, Please try again.')


    while True:
        try:
            free_cancelation = input('Free cancelation: ')
            if free_cancelation in ('Y','N'):
                break
            else:
                print('Wrong input, Please try again.')
        except:
             print('Wrong input, Please try again.')

    while True:
        try:
            Breakfast = input('Breakfast: ')
            if Breakfast in ('Y','N'):
                break
            else:
                print('Wrong input, Please try again.')
        except:
             print('Wrong input, Please try again.')
    
    return special_view,free_cancelation,Breakfast
```

```python
# get membership details.
def get_membership():
    while True:
        try:
            Membership = input('Member gold/silver/No (G/S/N): ')
            if Membership in ('G','S','N'):
                break
            else:
                print('Wrong input, Please try again.')
        except:
             print('Wrong input, Please try again.')
    return Membership
```

```python
roomtype = {
    1:'Room',
    2: 'Studio',
    3: 'One Bedroom Unit'
}

# define dictionary to hold room cost
# according to the room type

room_cost = {1:80,
            2:120,
            3:180}

room_cost_concession = {
    1:65,
    2:105,
    3:165
}


while True:
    
    print('WELCOME TO HOTEL BOOKINGS PLATFORM')
    print('Types of available rooms: \n  (1) Room \n  (2) Studio \n  (3) One Bedroom Unit')

    # call the function to get the inputs
    
    room_type = get_roomtype()

    no_of_guest = get_no_of_guest()

    no_of_days = get_no_of_day()

    no_of_concession = get_concession()

    special_view, free_cancelation, Breakfast =  get_additionals()

    Membership = get_membership()

    print()

    print('--- {} chosen ---'.format(roomtype[room_type]))

    print('--- You are booking for {} guests ---'.format(no_of_guest))
    print('--- {} days chosen --'.format(no_of_days))
    print('--- {} concessions chosen ---'.format(no_of_concession))

    print('--- {} ---'.format('Free cancelation chosen' if free_cancelation=='Y' else 'Free cancelation not chosen'))
    print('--- A room with {} chosen---'.format('Special view' if special_view=='Y' else 'Special view not'))
    print('--- {} chosen---'.format('Breakfast' if Breakfast=='Y' else 'Breakfast not'))

    member_ship = 'Gold' if Membership=='G' else ('Silver' if Membership=='S' else 'No')
    print('--- {} Membership discount applied---'.format(member_ship))
    print('--- Bulk discount {} applied ---'.format("has" if no_of_days>=6 else "hasn't"))

    # compute the room rent
    cost_without_concession = (no_of_guest-no_of_concession)*no_of_days*room_cost[room_type]
    cost_with_concession = no_of_concession*no_of_days*room_cost_concession[room_type]
    
    Total_Cost = cost_without_concession + cost_with_concession
    
    # compute charges of additional benefits
    Additional_percent_charge = (15 if special_view=='Y' else 0) + (3 if free_cancelation=='Y' else 0) + (3 if Breakfast=='Y' else 0)
    
    Total_Cost = Total_Cost + Total_Cost*Additional_percent_charge/100

    # percentage of discount
    Discount = (10 if no_of_days>6 else 0) + (10 if Membership=='G' else (5 if Membership=='S' else 0))
    total_discount = Total_Cost*Discount/100
    
    print()
    print('Final Ticket cost: {}$'.format(Total_Cost-total_discount))
    print('You saved in this transaction: {}$'.format(total_discount))

    while True:
        choice = input('Do you want to start a new booking? (Y)es or (N)o: ')
        if choice=='N':
            flag = False
            break
        elif choice!='Y':
            print('Wrong input, Please try again.')
        else:
            flag = True
            break
    print()
    if flag==False:
        break
```
