```python
# import library
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
```

# Q.2

```python
# generating random variable and normalising it to 0 to 1
# by dividing each element with maximum of x
x = norm.rvs(size=55,random_state = 10)
x = abs(x)/max(abs(x))

# generating random variable and normalising it to 0 to 1
# by dividing each element with maximum of y

y = norm.rvs(size=55,random_state = 5)
y = abs(y)/max(abs(y))
```

```python
# set the figure size
plt.figure(figsize=(8,5))
# draw scatter plt
plt.scatter(x,y,color='r',marker=None)
# set the title
plt.title('Scatter plot')
# set x and y label
plt.xlabel('x')
plt.ylabel('y')
# show the plot
plt.show()
```

# Q.3

```python
# define a function f(x) to find x*x + 15*np.sin(x) + np.tan(x)
def f(x):
    return x*x + 15*np.sin(x) + np.tan(x)
```

```python
# generate 500 evenly spaced numbers between -2 to 2
x = np.linspace(-2, 2, 500)
```

```python
y = f(x) # find value of f(x)
```

```python
# draw line chart
plt.plot(x,y,color='r')
# set the title
plt.title('Line chart')
# set x and y label
plt.xlabel('x')
plt.ylabel('f(x)')
# save the image
plt.savefig('linechart.png')
# show the plot
plt.show()
```
