```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x,y): # sin(sqrt(x**2+y**2))
    return np.sin(np.sqrt(x ** 2 + y ** 2))
```

```python
x = np.linspace(-10, 10, 40)
y = np.linspace(-10, 10, 40)
```

```python
X,Y = np.meshgrid(x, y)
Z = f(X, Y)
```

# 3D Contour Plots

```python
fig = plt.figure(figsize=(7,7))
axis = plt.axes(projection='3d')
axis.contour3D(X, Y, Z, 60, cmap='plasma')
axis.set_xlabel('x')
axis.set_ylabel('y')
axis.set_zlabel('z');
```

```python
# changing the viewing angle i.e 70 degree about x-y plane and rotated 25 degree counter clockwise about z axis
axis.view_init(70, 25) 
fig
```

# Wire frame plot

```python
fig = plt.figure(figsize=(7,7))
axis = plt.axes(projection='3d')
axis.plot_wireframe(X, Y, Z, color='red')
axis.set_title('Wireframe plot');
```

# surface plot

```python
fig = plt.figure(figsize=(7,7))
axis = plt.axes(projection='3d')
axis.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='plasma', edgecolor='red')
axis.set_title('surface plot');
```
