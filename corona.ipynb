```python
# import library
import pandas as pd
```

```python
# source of dataset 
# https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/05-10-2021.csv
rdf = pd.read_csv('covid_data.csv')
```

```python
rdf = rdf.iloc[:,3:] # consider only important columns
```

```python
rdf.head(5) # display top 5 rows
```

```python
# create pivot table
country_deaths = rdf.pivot_table('Deaths',index='Country_Region').sort_values(by='Deaths',ascending=False)
country_deaths
```

# display all rows

```python
len(list(country_deaths))
```

```python
country_deaths.shape[0]
```

```python
# set number of rows to be displayed
pd.set_option('display.max_rows',len(country_deaths))
# set number of columns to be displayed
pd.set_option('display.max_columns',len(list(country_deaths)))

country_deaths
```

```python
# reset the number of rows and columns to their default value
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
```

# Plot

```python
import matplotlib.pyplot as plt
```

```python
totals = country_deaths.groupby('Country_Region')["Deaths"].sum().sort_values()
totals
```

```python
plt.figure(figsize=(5,50)) # set the figure size
totals.plot(kind='barh',fontsize=10,color='r')
plt.xticks(rotation='vertical')
plt.show() # plot the graph
```
