```python
import os,time
os.environ['TZ']='South America/Argentina'                                            
time.tzset()                                                             
 
from datetime import date,time,datetime
today = date.today()                                                     
today_start = datetime(today.year, today.month, today.day, 0, 0, 0)      
today_float = datetime.timestamp(today_start)                            
now = datetime.now()                                                    
current_time = time(now.hour, now.minute, now.second)                    
time_now = datetime.combine(today, current_time)                         
time_now_float = time_now.timestamp()                                    
tnf = time_now_float                                          
 
## testing na info lang
name = ["Name1", "Name2", "Name3", "Name4", "Name5"]
address = ["Address1","Address2","Address3","Address4","Address5"]
contact = ["Contact1","Contact2","Contact3","Contact4","Contact5"]
temp = ["37.5","37.7","37.9","38.1","38.3",]
time_enter = []
time_leave = []
 
##time1 in and out
time1 = datetime(2021,5,today.day - 1,6,0,0)   
time1_float = time1.timestamp()
time_enter.append(time1_float)
time1out = datetime(2021,5,today.day - 1,7,0,0) 
time1out_float = time1out.timestamp()
time_leave.append(time1out_float)
 
##time2 in and out
time2 = datetime(2021,5,today.day - 1,8,0,0)   
time2_float = time2.timestamp()
time_enter.append(time2_float)
time2out = datetime(2021,5,today.day - 1,9,0,0) 
time2out_float = time2out.timestamp()
time_leave.append(time1out_float)
 
##time3 in and out
time3 = datetime(2021,5,today.day,6,0,0)   
time3_float = time3.timestamp()
time_enter.append(time3_float)
time3out = datetime(2021,5,today.day,8,0,0) 
time3out_float = time3out.timestamp()
time_leave.append(time3out_float)
 
##time4
time4 = datetime(2021,5,today.day,7,0,0)   
time4_float = time4.timestamp()
time_enter.append(time4_float)
time_leave.append(0)
 
##time5
time5 = datetime(2021,5,today.day,8,0,0)   
time5_float = time5.timestamp()
time_enter.append(time5_float)
time_leave.append(0)
 
currently_in = 0                     
for x in time_leave:
    if x == 0:
        currently_in += 1
print("-------------------------------------------")
print("            Welcome to ________!           ")
print("  There are currently {} person/s inside. ".format(currently_in))
print("\n[1] Enter information for contact tracing\n[2] Show customers for a certain day\n[3] Search for a specific person\n[4] Time-out to leave the building\n[5] Exit")
print("-------------------------------------------")
start_input = input("Input: ")
 
while start_input != "1" and start_input != "2" and start_input != "3" and start_input != "4" and start_input != "5":      
    print("Please enter the corresponding number for each option.")
    start_input = input("Input: ")
while True:
    if start_input == "1":                                       
        print("-------------------------------------------")
        print("   Please enter the required details for")
        print("        possible contact tracing.")
        print("-------------------------------------------")
        name_in = input("Full Name: ")
        name.append(name_in)
        address_in = input("Complete Address: ")
        address.append(address_in)
        contact_in = input("Contact Number: ")
        contact.append(contact_in)
        temp_in = input("Temperature (Celsius): ")
        temp.append(temp_in)
        time_enter.append(tnf)
        time_leave.append(0)
        print("-------------------------------------------")
        print("    Successfully entered contact info!")
        print("-------------------------------------------")

    elif start_input == "2":
        print("-------------------------------------------")
        print("     Please enter the specific date.\n")
        print("   For the day, enter a value from 1-31")
        print("  For the month, enter a value from 1-12")
        print("-------------------------------------------")
        search_day = int(input("Day: "))
        search_month = int(input("Month: "))
        search_year = int(input("Year: "))

        search = date(search_year, search_month, search_day)                          
        search_date = datetime(search.year, search.month, search.day, 0,0,0)            
        search_date_limit = datetime(search.year, search.month, search.day+1,0,0,0)     
        search_date_float = search_date.timestamp()                                    
        search_date_limit_float = search_date_limit.timestamp()                         

        print("-------------------------------------------")
        print("Seaching for entries on the date {}".format(search))
        counter = 0
        for x in time_enter:
            if x >= search_date_float and x <= search_date_limit_float:                  
                i = time_enter.index(x)                                                     
                if time_leave[i] == 0:
                    print("-------------------------------------------")
                    print("Name:", name[i], "\nAddress:", address[i], "\nContact Number:", contact[i], "\nTemperature:", temp[i], "\nTime In:", datetime.fromtimestamp(time_enter[i]), "\nTime Out: Still inside")   
                    counter += 1                                                              
                else:
                    print("-------------------------------------------")
                    print("Name:", name[i], "\nAddress:", address[i], "\nContact Number:", contact[i], "\nTemperature:", temp[i], "\nTime In:", datetime.fromtimestamp(time_enter[i]), "\nTime Out:", datetime.fromtimestamp(time_leave[i]))     
                    counter += 1
        print("-------------------------------------------")
        print("   End of results. Number of results: {}".format(counter))
        print("-------------------------------------------")

    elif start_input == "3":
        print("-------------------------------------------")
        print("    Please enter the name of the person")
        print("         you want to search for.")
        print("            (Case Sensitive).")
        print("-------------------------------------------")
        person = input("Search for Name: ")
        for y in name:                                                                 
            if y.__contains__(person):             
                ind = name.index(y)
                print("-------------------------------------------")        
                print("Name:", name[ind], "\nAddress:", address[ind], "\nContact Number:", contact[ind], "\nTemperature:", temp[ind], "\nTime In:", datetime.fromtimestamp(time_enter[ind]), "\nTime Out:", datetime.fromtimestamp(time_leave[ind]))      #prints the corresponding info galing sa list ng names, etc.
        print("-------------------------------------------")
        print("              End of results")
        print("-------------------------------------------")
 
    elif start_input == "4":
        print("-------------------------------------------")
        print("    You are now leaving the building.\n")
        print("   Please enter your name to time-out.")
        print("            (Case Sensitive).")
        print("-------------------------------------------")
        name_leave = input("Name:")
        if any(name_leave in s for s in name):            
            for x in name:
                if x.__contains__(name_leave):
                    i = name.index(x)                          
                    if time_leave[i] == 0:                      
                        print("-------------------------------------------")
                        print("       Is this your contact info?")
                        print("-------------------------------------------")
                        print("Name:", name[i], "\nAddress:", address[i], "\nContact Number:", contact[i], "\nTemperature:", temp[i], "\nTime In:", datetime.fromtimestamp(time_enter[i]))
                        print("\n       [1] Yes            [2] No")
                        print("-------------------------------------------")
                        leave_confirm = input("Input:")
                        if leave_confirm == "1":                  
                            print("-------------------------------------------")
                            print("         Timed-out Successfully!")
                            print("-------------------------------------------")
                            time_leave[i] = time_now_float
                        else:
                            print("-------------------------------------------")
                            print("Cannot find name/Person is currently not in")
                            print("   Please check for misspellings/errors.")
                            print("-------------------------------------------")
                    else:
                        print("-------------------------------------------")
                        print("Cannot find name/Person is currently not in")
                        print("   Please check for misspellings/errors.")
                        print("-------------------------------------------")
                else:
                    print("-------------------------------------------")
                    print("Cannot find name/Person is currently not in")
                    print("   Please check for misspellings/errors.")
                    print("-------------------------------------------")

        
    else:                                
        break
  
    currently_in = 0                      
    for x in time_leave:
        if x == 0:
            currently_in += 1
 
    print("-------------------------------------------")
    print("            Welcome to ________!           ")
    print("  There are currently {} person/s inside. ".format(currently_in))
    print("\n[1] Enter information for contact tracing\n[2] Show customers for a certain day\n[3] Search for a specific person\n[4] Time-out to leave the building\n[5] Exit")
    print("-------------------------------------------")
    start_input = input("Input: ")
 
    print("-------------------------------------------")
    print("          THANK YOU")
    print("-------------------------------------------")
```
