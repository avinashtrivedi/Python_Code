```python
# import the library
import xml.etree.ElementTree as ET

def main():
    year = input("Please enter a year: ")
    cod_list = []
    
    # parse the xml file
    tree = ET.parse('leading_causes_of_death.xml')
    
    # get the root
    root = tree.getroot()
    
    # iterate through the xml rows
    for row in root[0]:
        
        # call the process_row() function to get the row as dictionary
        cod_list.append(process_row(row))
        
    # call the print_rows() function 
    print_rows(cod_list,year)
```

```python
def print_rows(row_list,year):
    for indx,record in enumerate(row_list):
        if record['year']==year:
            print("Record {} (id: {}): During {} the leading cause of death in {} was {}".
                  format(indx,record['_id'],year,record['state'],record['cause_name']))
```

```python
def process_row(row):
    
    # dictionary to store key: value pairs
    mydict = {}
    mydict['_id'] = row.get('_id')
    
    # iterate through each elements
    for subelem in row:
        
        # create a dictionary items
        mydict[subelem.tag] = subelem.text
        
    return mydict
```

```python
if __name__=='__main__':
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import LineString

x = np.arange(0, 1000)
f = np.arange(0, 1000)
g = np.sin(np.arange(0, 10, 0.01) * 2) * 1000

plt.plot(x, f)
plt.plot(x, g)

first_line = LineString(np.column_stack((x, f)))
second_line = LineString(np.column_stack((x, g)))
intersection = first_line.intersection(second_line)

if intersection.geom_type == 'MultiPoint':
    plt.plot(*LineString(intersection).xy, 'o')
elif intersection.geom_type == 'Point':
    plt.plot(*intersection.xy, 'o')
```
