```python
# import the library
import pandas as pd
```

```python
# load the dataset
df = pd.read_csv('IthacaDailyClimate2018.csv')
```

```python
# an expression to extract a new dataframe containing those days 
# where the temperature reached at least 70 degrees
at_least_70 = df[df['Maximum Temperature']>=70]

# expression that computes how many days reached at least 70 degrees
num_at_least_70 = len(at_least_70)
```
