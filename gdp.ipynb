```python
def open_file():
    '''Repeatedly prompt until a valid file name allows the file to be opened.'''
    while True: # prompt for a valid file until file is properly opened
        try: 
            file = input('Enter a file name: ')
            fp = open(file)  # open a file
            return fp # return file pointer
        except OSError as e:
            print('Error. Please try again')
    
def find_min_percent(line):
    '''Find the min percent change in the line; return the value and the index.'''
    min_value = 10000000
    min_value_index = 0
    # As the number of data items are 47
    # this loop will run from 1 to 47 and create start and end index for 
    # each data items
    for i in range(1,48,1): 
        # start and end index for each data items
        start = 76 + (i-1)*12
        end = 76 + i*12
        
        # strip() : remove leading and trailing whitespace from a string
        # float() : convert string value to float
        value = float(line[start:end].strip()) 
        if value < min_value:
            min_value = value
            min_value_index  = start
    return min_value,min_value_index

def find_max_percent(line):
    '''Find the max percent change in the line; return the value and the index.'''
    max_value = -10000000
    max_value_index= 0
    for i in range(1,48,1):
        # start and end index for each data items
        start = 76 + (i-1)*12
        end = 76 + i*12
        
        # strip() : remove leading and trailing whitespace from a string
        # float() : convert string value to float
        value = float(line[start:end].strip())
        if value > max_value:
            max_value = value
            max_value_index = start
    return max_value,max_value_index

def find_gdp(line, index):
    '''Use the index fo find the gdp value in the line; return the value'''
    return float(line[index:index+12].strip())

        
def display(min_val, min_year, min_val_gdp, max_val, max_year, max_val_gdp):
    '''Display values; convert billions to trillions first.''' 
    print() # print blank line
    # trillion = billion/1000
    print("{:<10}{:>8}{:>6}{:>18}".format('min/max','change','year','GDP (trillions)'))
    print("{:<10s}{:>8.1f}{:>6d}{:>18.2f}".format('min',min_val,min_year,min_val_gdp/1000))
    print("{:<10s}{:>8.1f}{:>6d}{:>18.2f}".format('max',max_val,max_year,max_val_gdp/1000))

def main():                    
    line9 ='' # for line 9 , GDP Change
    line44='' # for line 44 , Actual GDP
    year=''   # for year
    fp = open_file() # call this function to open a file
    
    # As list,tuple etc.. are not allowed
    # Hence,reading line as a string
    # This loop will run for 44 times so that it can read line 9, line 44 
    # and line 8 i.e. Year
    for i in range(44):
        temp = fp.readline()
        if i+1 == 9:
            line9 = temp
        elif i+1==44:
            line44 = temp
        elif i+1==8:
            year = temp
            
    # get min and max annual change in GDP and corresponding indexes
    min_val,index_min = find_min_percent(line9)
    max_val,index_max = find_max_percent(line9)
    
    # setting the values so as to call display() function
    
    min_val = min_val
    min_year = int(year[index_min:index_min+12].strip())
    min_val_gdp = find_gdp(line44,index_min)
    
    max_val = max_val
    max_year = int(year[index_max:index_max+12].strip())
    max_val_gdp = find_gdp(line44,index_max)
    
    display(min_val, min_year, min_val_gdp, max_val, max_year, max_val_gdp)
    
# Calls the 'main' function only when you execute within Spyder (or console)
# Do not modify the next two lines.
if __name__ == "__main__":
    main()
```
